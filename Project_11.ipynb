## Spatial solitons in periodic nanostructures review

the existence and stability of TE and TM spatial solitons in a subwavelength periodic semiconductor-dielectric structure. We have found that for the wavelength of 1550 nm and the interface separation close to and less than 100 nm the band structure of the linear TE waves becomes similar to the band structure of a homogeneous medium. The properties of TE solitons change
accordingly

In this work, we consider an infinite array of semiconductor photonic wires with the wire widths and separations taken
well below the wavelength. We assume that the photon energy is below the band gap and the semiconductor acts essentially 
as a high-index nonlinear dielectric. We present analysis of linear and soliton solutions for TE polarizations
using first-principles Maxwell equations. The TE solitons undergo a smooth transformation from the case,
which can be well understood in terms of the evanescently coupled waveguides into the solitons of a quasihomogeneous
medium, when the widths of all layers become much less than the wavelength.

The array of slot waveguides we consider below is a periodic structure of narrow layers of a high-index semiconductor material 
(e.g., silicon) embedded into a lowindex dielectric material (e.g., silica glass).  The separations of the semiconductor
layers vary from 500 to 50 nm, and the width of the layers is
between 220 and 95 nm. The vacuum wavelength is taken to
be vac= 1.55 m. Our analysis is based on the nonlinear
Maxwell equations in the two-dimensional geometry

$\vec{\nabla} X \vec{H} = -ikc\epsilon_{0}\vec{D},  \hspace{1cm}   \vec{\nabla} X \vec{E} = \frac{-ik}{c\epsilon_{0}}\vec{H}$

 where $k= 2/\lambda_{vac}$, c is the speed of light in vacuum, $\epsilon_{0}$ is the
vacuum permittivity, and for electric $\textit{E}$ ,  and magnetic $\textit{H}$ fields
it is assumed that $\textit{E} , \textit{H}$  = $\frac{1}{2}$ E, H.exp{−ikct}+c. Light propagation is assumed to be along the z direction and the x coordinate is perpendicular to the layers 

For the TE-polarized modes, only $E_y$, $H_x$, and $H_z$ are
nonzero and the resulting equations are

$\partial_{xx}E_{y} =-k^{2}D_{y}$

$\frac{k}{c\epsilon_{0}}H_{x}=i\partial_zE_{y}$--(1) $\hspace{1cm}$       $\frac{k}{c\epsilon_{0}}H_{z}=i\partial_xE_{y}$ --(2)

We only need to solve these equations for the electric field
components in both the TE and TM cases. The constitutive
relation is taken as for isotropic materials

$ \vec{D}=\epsilon\vec{E} = \frac{1}{2}\chi_3[|\vec{E}|^2\vec{E} + \frac{1}{2}(\vec{E}.\vec{E})\vec{E}^*]$

we do not force boundary conditions at the interfaces but instead assume that the linear
permittivity  and the nonlinear susceptibility 3 change continuously but sharply between their respective values for
the material s silicon and material g silica glass

$\epsilon(x)= \epsilon_g +  \sum_{j} (\epsilon_s-\epsilon_g)K_j(x) $

where $K_j = exp{([-(x−x_j)/w]^{10})}	$,    $\epsilon=n^2, n_g= 1.44,$ and $n_s= 3.48$

Soliton solutions of Eqs. 1 and 2 are sought in the
form

$E_y = u(x)exp(iqkz) $

which results in the familiar stationary nonlinear Schrödinger
equation NLSE,


$ u^{''} + k^2(\epsilon − q^2)u + k^2\gamma^3 =0  $. --(3)

In a case of the evanescently coupled semiconductor
waveguides, the above equation is readily transformed into a
set of coupled mode algebraic equations, giving familiar discrete soliton solutions 10. In the limit when w and s are
much less then $\lambda_{vac}$, one should expect the soliton profiles to
be close to the ones known from the NLSE with constant
coefficients

### LINEAR MODES, BAND STRUCTURE

For $\gamma=0$ Eq.3 becomes

$ u^{''} + k^2\epsilon = q^2u   $

jumps in $\epsilon$ force the second derivative of u to

we choose w= 220 nm and a sufficiently large separation between the semiconductor layers s= 500 nm and solve for the eigen vectors u(x) and eigen values $q^2$

The $k_B$= 0 and $k_B =\pi/k(w+s) $

```python
import numpy as np
import scipy
from findiff import FinDiff
import matplotlib.pyplot as plt
```

```python
pi = np.pi

wavelength = 5    # We take 100 nanometres to be 1 unit, so it means wavelength of the light considered is 500nm
k = 2*pi/wavelength
n_g = 1.44   # refractive index of glass
epsilon_g = n_g**2
n_s = 3.48     # refractive index of semiconductor used
epsilon_s = n_s**2
```

```python
def K(x,j,s,w):           # defining a supergaussian function
    y = (-1)*((x-j*(w+s))/w)**10
    return np.exp(y)
```

```python
def epsilon(x,w,s):          # permittivity as a function of x

    sum_of_K_s = 0
    for i in range(-100,100):
        sum_of_K_s += K(x,i,s,w)
        
    return epsilon_g + (epsilon_s-epsilon_g)*sum_of_K_s
```

```python
def TE_linear(w,s,kb, plot_or_not):    # this function solves the eigenvalue equation for linear mode of TE waves

  nx = 101
  x = np.linspace(-10, 10, nx)
  dx = x[1] - x[0]

  d2_dx2 = FinDiff(0, dx, 2, acc=4).matrix(x.shape)    # matrix form approximation of double derivative operator
  d_dx = FinDiff(0, dx, 1, acc=4).matrix(x.shape)    # matrix form approximation of derivative operator

  eps = scipy.sparse.diags(epsilon(x,w,s))       # converting epsilon to a 2D matrix

  arr = np.eye(nx)
  kb_square = (kb**2)*arr

  kb_square = scipy.sparse.csr_matrix(arr)    # kb_square is a diagonal 2D matrix, with all diagonal entries being kb**2

  H = d2_dx2 + (k**2)*(eps-kb_square) + 2*(1j)*k*kb*d_dx    # final operator, for which we want to solve eigenvalue equation
  
  spectrum, states = scipy.sparse.linalg.eigs(H, which='SR', k=1)   # eigenvalue equation

  # eigenfunctions derived till now represent r(x)
  # Now, we calculate u(x) from r(x)

  states = states*(np.exp((1j)*k*kb))
  states = states*x  # representation for u(x)

  # plotting the eigenstate
  
  if plot_or_not:
    for n in range(1):
        plt.plot(x, states[:, n], '.-')

    plt.xlim(0,10)
    plt.xlabel('x')
    plt.ylabel('u')
    plt.grid()

  return spectrum, states
```

```python
def plot_eigenvalues(w,s):  # function to plot eigenvalues for a range of kb values

  lower_kb = -3/(k*(w+s))
  upper_kb = 3/(k*(w+s))
  step_kb = 0.1/(k*(w+s))

  number_of_kb = (upper_kb - lower_kb)/step_kb+1
  number_of_kb = int(number_of_kb)

  # defining zero arrays
  q_square_values = np.zeros(number_of_kb)  # to store eigenvalues
  x_axis_values = np.zeros(number_of_kb)    # to store x-axis coordinates


  for i in range(0, number_of_kb):

    kb = lower_kb + i*step_kb

    spectrum, states = TE_linear(2.2, 5, kb, False)   # store eigenvalues in 'q_square_values' array
    spectrum = spectrum[0]
    q_square_values[i] = spectrum

    x_axis_values[i] = kb*k*(w+s)


  plt.plot(x_axis_values, q_square_values)
  plt.show
```

```python
plot_eigenvalues(2.2,5)    # plot of q^2 when w = 220nm, s = 500nm
```

```python
_,_ = TE_linear(2.2, 5, 0, True)    # plot of u(x) with w = 220nm, s = 500nm and kb = 0
```

```python
_, _ = TE_linear(2.2, 5, pi/(k*(2.2+5)), True)    # plot of u(x) with w = 220nm, s = 500nm and kb = pi/k(w+s)
```
