# Abstract/Introduction
The paper describes the enhancement of non-linear response of a resonant and atomically thin material by placing it in front of a partially reflecting mirror. The non-linear response is obtained by the long lived polaritons that exists only for specific distances between the thin material and the mirror. This scheme eliminates the need for cavities and complex nanophotonic structures. Finally, a specific implementation on exciton-polariton resonances in two-dimensional semiconductors and the role of imperfections and loss is discussed.

Before the paper can be discussed, it is worthy to take note of a few key words and concepts, while some key words are explained as the paper is being discussed.

**Exciton**: An exciton is a bound state of an electron and an electron hole which are attracted to each other by the electrostatic Coulomb force.

**Polariton**: Polaritons are quasiparticles resulting from strong coupling of electromagnetic waves with an electric or magnetic dipole-carrying excitation. They are an expression of the common quantum phenomenon known as level repulsion, in essense they are superposition of exciton and photon.


# QUANTUM NON-LINEAR OPTICS IN ATOMICALLY THIN MATERIALS

This paper describes a novel approach to Non-Linearity in Optical response which uses resonant and atomically thin material (2-dimensional monolayer of TMD-Transition Metal Dichalcogenides). 

The speciality of TMD is that it supports tightly bound, optically active excitons. It is obsearved that excitons with zero in-plane momentum decay by emitting a plane wave on the either side of the TMD. To enhance the non-linear response, the TMD monolayer is placed in front of a partially reflecting mirror.

Untitled2.png

This setup can be understood as an atom replacing the two-dimensional emitter and a one-dimensional waveguide replacing the role of a free space plane wave.

The excitons have energy levels similar to that of harmonic oscillator in the absense of non-linearity (Hamiltonian is in the form of $a^\dagger$$a$) and when the excitons are shined with light, they get excited to higher energy levels and the resonance frequency is $\omega_0$. Then the excitons decay by emitting light whose frequency is also $\omega_0$ with a rate of $\gamma$. However there is a finite linewidth to this decay which is also equal to $\gamma$ (in essense, the emitted light will contain wavelengths within the linewidth around the resonant wavelength). There are other forms of decay apart from emission of photons, which can be the emission of phonons(vibrational modes) and non-radiative decay which occurs due to material imperfections. The difference between the fequency of the incident light and the resonant frequency is called detuning(detoned by $\delta$)

The excitons emit a plane wave on either side of the TMD with a radiative rate $\gamma$ and a loss rate $\gamma\prime$. If the seperation between the TMD and the mirror is an integral multiple of half of the exciton resonant wavelength, the light emitted towards the left and the light reflected by the mirror destructively interfere, which leads to suppression in the radiative linewidth, enhancement of the exciton lifetime and also the optical nonlinearity.

It leads to the formation of long lived exciton-polaritons, where the excitonic component is predominant. The excitonic component is very sensitive to non-linear frequency shifts arising from exciton-exciton interaction.

Before proceeding to the non-linear case, it is worth examining the linear response of a TMD in free space. Close to excitonic resonance, the amplitude reflection coefficient is given by the Lorentzian $r_{\mathrm{TMD}}(\delta)=-i(\gamma / 2) /\left[\delta+i\left(\gamma+\gamma^{\prime}\right) / 2\right]$

where $\delta$ is the detuning from the resonance. Even though the TMD layer is much thinner than the resonance wavelength, it acts like a perfect reflector at zero detuning and no losses. The difference between this case and the Fabry-Perot resonator is that in the case of FPR, the reflection coefficient is independent of incident frequency.

The intensity transmission coefficent of the TMD can be computed by summing over all multiple reflections between the TMD and the mirror. The following pictures reveal that the transmission spectrum strongly depends on the distance between TMD and the mirror.

# Fabry-Perot Resonator

  An optical resonator is an arrangement of mirrors that support the formation of a standing wave, it is the counterpart of electronic resonant circuit, which confines and stores light at certain frequencies. Fabry-Perot resonator is a case where there are two plane mirrors(with finite reflectance and transmission coefficients) parallel to each other seperated by a distance ‘L’.

Consider the simpler case where the mirrors have zero transmission coefficient and light is confined between them. The optical length between the two mirrors should be an integral multiple of half of the wavelength of the light confined between the mirrors.
$\lambda_{q}=\frac{2 L}{q}$;
$\lambda_q$  is the wavelength of light between mirrors and q is a natural number.

Thus the frequencies supported by a resonator have a spacing of $\Delta v=\frac{c_{0}}{\lambda_{q}}-\frac{c_{0}}{\lambda_{q^{+1}}}=\frac{c_{0}}{2 L}$.


Untitled.png

If the two mirrors have finite transmission coefficients, the intensity peaks broaden and other frequencies which do not follow the previous relation will induce a resonant behaviour inside the resonator. And the losses like scattering and absorption at mirror surfaces can be incorporated by placing a medium with loss factor 'V'.

Monochromatic light of wavelength $\lambda$ of intensity $I_0$(Electric field $E_0$) will be partially reflected by mirror 1. The transmitted intensity portion is then reflected back and forth between the two mirrors. With each reflection intensity is coupled out of the Fabry-Perot Resonator in both the forward and the backward direction. Summation of all waves coupled out yields the transmission T and the reflectance R of the Fabry-Perot Resonator. If $r_1$, $r_2$, denote the reflection coefficients and $t_1$,$t_2$ are the corresponding transmission coefficients of the two mirrors, the reflected field $E_r$ and the transmitted field $E_t$ are given by  (with V being the amplitude loss factor):
\begin{array}{l}
E_{r}=E_{0}\left[r_{1}-t_{1}^{2} v^{2} r_{2} \exp [i 2 k L] \sum_{n=0}^{\infty}\left(v^{2} r_{1} r_{2} \exp [i 2 k L]\right)^{n}\right] \\
E_{t}=-E_{0} t_{1} t_{2} v \exp [i k L] \sum_{n=0}^{\infty}\left(v^{2} r_{1} r_{2} \exp [i 2 k L]\right)^{n}
\end{array}
Note that a transmission phase factor of $exp(i\pi/2)$ has to be taken into account to get the minus sign

Untitled1.png

By calculating the intensities, we can determine the fraction of the incident intensity $I_0$ being transmitted and reflected by Fabry-Perot resonator: 
\begin{array}{l}
R=\frac{\left(\sqrt{R_{1}}-\sqrt{R_{2}} V\right)^{2}+4 \sqrt{R_{1} R_{2}} V \sin ^{2}(k L)}{\left(1-\sqrt{R_{1} R_{2}} V\right)^{2}+4 \sqrt{R_{1} R_{2}} V \sin ^{2}(k L)} \\
T=\frac{\left(1-R_{1}\right)\left(1-R_{2}\right) V}{\left(1-\sqrt{R_{1} R_{2}} V\right)^{2}+4 \sqrt{R_{1} R_{2}} V \sin ^{2}(k L)}
\end{array}
with $R_i$=$\mid r_i\mid^2$: reflectivity of mirror i,  $V=\mid v\mid^2$: loss factor per transit, k= $2\pi/\lambda$: wave number, $\lambda$ : wavelength in vaccum and L : optical mirror spacing.

The transmission T becomes maximum for $sin(kL)=0$ $\implies$ $\lambda=(2L/q)$ :    $q=1,2,3...$

As we saw above, perfect transmission through a fabry perot resonator occurs when :

  

*   The reflection cooficients of the two mirrors are equal
*   The round trip phase accumulated in the resonator is an integer multiple of 2$\pi$.

Some important variables and parameters:

- $r_0 , t_0$ : amplitude reflection and transmission cooficient of the mirror places in front of the TMD.
- $R_0 , T_0$ : Intensity reflection and transmission cooficient of the mirror placed in front of the TMD.
- $r(\delta), t(\delta)$ : amplitude reflection and transmission cooficient of the mirror+ TMD system together
- $R(\delta), T(\delta)$ : Intensity reflection and transmission cooficient of the mirror+ TMD system together
- $r_{TMD}(\delta)$ : ampitude reflection cooficient of the TMD.


To satisfy the first condition, we need $\left|r_{\mathrm{TMD}}(\delta)\right|^{2}=\left|r_{0}\right|^{2}$, where $r_{TMD}(\delta)$ is known to be given by: $-i(\gamma / 2) /\left[\delta+i\left(\gamma+\gamma^{\prime}\right) / 2\right]$. Here $r_0$ is the reflectance of the mirror and is given by a negative value.

This will give us the detuning value at which we observe perfect transition. Solving the equation:
\begin{equation}
|-i(\gamma / 2) /\left[\delta+i\left(\gamma+\gamma^{\prime}\right) / 2\right]|^{2}=\left|r_{0}\right|^{2}
\end{equation} \\
Considering a scenario where the non radiative decay rate $\gamma^{\prime}$ is zero, this gives  $\delta_{\mathrm{FP}}=\pm(\gamma / 2) \sqrt{T_{0} / R_{0}}$ where $T_0$ and $R_0$ are the intensity transmission and reflection cooficients of the mirror respectively, which are given by $T_0 = |t_0|^{2}$ and $R_0 = |r_0|^{2}$. We consider the case of a highly reflecting mirror such that the intensity transmission cooficient $T_0$ is very small.

We have another parameter which decides whether we have perfect transmission of not. The parameter which decides this is *$d$* (the distance between the mirror and the TMD)  just like for the fabry perot resonator scenario.

Let's find the reflection and transmission of the entire system(mirror and the TMD) by accounting for the multiple reflections between them and also the finite transmissions in each steps. Summing over all such processes will give us a net cooficient for the combined system given by:
\begin{equation}
\begin{array}{c}
r(\delta)=r_{\mathrm{TMD}}(\delta)+t_{\mathrm{TMD}}(\delta)^{2} r_{0} e^{2 i k d}\left[1+r_{\mathrm{TMD}}(\delta) r_{0} e^{2 i k d}+r_{\mathrm{TMD}}(\delta)^{2} r_{0}^{2} e^{4 i k d}+\cdots\right] \\
\\
t(\delta)=t_{\mathrm{TMD}}(\delta)\left[1+r_{\mathrm{TMD}}(\delta) r_{0} e^{2 i k d}+r_{\mathrm{TMD}}(\delta)^{2} r_{0}^{2} e^{4 i k d}+\cdots\right] t_{0} e^{i k d}
\end{array}
\end{equation}

Which is quite evident if we look the picture below.





Screenshot 2020-10-30 164653.jpg

The above two equations are simple geometric progressions and can be simplified using the formula for sum of an infinite geometric progression.
They reduce to:
\begin{equation}
\begin{array}{c}
r(\delta)=r_{0} e^{2 i k d}-\frac{i \gamma / 2\left(1+r_{0} e^{2 i k d}\right)^{2}}{\delta+i \gamma / 2\left(1+r_{0} e^{2 i k d}\right)+i \gamma^{\prime} / 2} \\
\\
t(\delta)=\frac{\delta+i \gamma^{\prime} / 2}{\delta+i \gamma / 2\left(1+r_{0} e^{2 i k d}\right)+i \gamma^{\prime} / 2} t_{0} e^{i k d}
\end{array}
\end{equation}

\\
For the condition where $\gamma^{\prime}$ is $0$, we get the condition for perfect transmission that is $|t(\delta)|^2=1$ when the value of $\delta$=$\delta_{FP}$ and $r_{0} e^{2 i k_{\mathrm{FP}} d_{\mathrm{FP}}}=-R_{0} \pm i \sqrt{R_{0} T_{0}}$. This second relation gives us the separation for which we get perfect transmission. The value of $d$ given by this is known as $d_{FP}$ and exactly satisfies the second condition required by a fabry perot resonator that is the phase accumulated in a round trip = $2\pi$.

The code below plots the intensity transmission cooficient $T$ as a function of detuning ($\delta$) for different values of $d$. Non radiative decay rate $\gamma^{\prime}$ has been taken to be 0 for this plot. We can clearly see that for $d = d_{FP}$ and $\delta=\delta_{FP}$ we see a sharp peak and the $T$ becomes 1 which signifies perfect transmission. We also see that for all values of $d$, at detuning $\delta=0$ we get $T=0$ which is as expected since the TMD acts like a perfect reflector for the case of 0 detuning.

```python
f = lambda x, d: (0.1*(x*x))/(((x+0.5*np.sqrt(0.9)*(np.sin(4*np.pi*d)))**2)+0.25*((1-np.sqrt(0.9)*np.cos(4*np.pi*d))**2))
x = np.linspace(-3, 3, 400)
a = complex(0,-(0.25/np.pi)*np.log(np.sqrt(0.9)-complex(0,np.sqrt(0.1))))
d = [0.25,3.0/8.0, 0.5,a]
lab=["$d=2\lambda_{0}/8$","$d=3\lambda_{0}/8$","$d=4\lambda_{0}/8$","$d=d_{FP}$"]

for i in range(4): 
    y = [f(j, d[i]) for j in x] 
    plt.plot(x, y,label=lab[i])
    plt.ylim(0, 1)



plt.grid()

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',prop={'size': 15}, borderaxespad=0.)
plt.ylabel("T")
plt.xlabel(r'$\delta/\gamma$')
plt.show()
```

An important feature to note in the above plot is that the radiative linewidth for excitonic resonance (linewidth when $\delta=0$) for the case where $d=\lambda/2$ is much less than that for the cases $3\lambda/8$ and $2\lambda/8$. This is due to the fact that the presence of the mirror at $d$ values close to $\lambda/2$, the field emmited by the TMD and the light reflected from the mirror destructively interfere, thus reducing the radiative linewidth which in turn leads to extended lifetimes of the exciton.


Now, we take a look at the fabry perot resonance linewidth $\gamma_{FP}$. This is the linewidth of the peak of the fabry perot resonance. It is important to analyse this quantity since it gives the lifetime of the polariton at the fabry perot resonance frequency and also plays a significant role in the non linear features of the system which we discuss later. 

We can calculate $\gamma_{FP}$ by taking into account the phase accumulated during the course of multiple reflections between the TMD and the mirror. We saw the we have infinite series of terms in the equation for reflection cooficient with each next term getting subsequently smaller in magnitude. For a small $T_0$ we consider the contribution due to the first $N=1/T_0$ round trips between the TMD and the mirror.

If a photon is detuned by an amount $\Delta$ from the fabry perot resonance frequency, then it acquires an additional phase during these round trips given by the value $\varphi_{\text {prop }}(\Delta)=2 N d \Delta / c$, where $c$ is the speed of light.
It also requires a phase due to reflections from the TMD which is given by: $\varphi_{\mathrm{TMD}}(\Delta) \approx  N \Delta / \gamma$.

The width of resonance can be calculated from the fact that the net phase acquired at detuning= linewidth($\gamma_{FP}$) is equal to 1.
Thus:
\begin{equation}
 \varphi_{\text {prop }}(\gamma_{FP})+\varphi_{\mathrm{TMD}}(\gamma_{FP})\approx 1
\end{equation}
We work in the limit where $\varphi_{\mathrm{TMD}}>>\varphi_{\text {prop }}$. This condition imposes that $N \Delta / \gamma>>2 N d \Delta / c$ which simplifies to $\gamma<<c/2d$ (also called free spectral range). 



Using the above approximation, we get $\varphi_{\mathrm{TMD}}(\gamma_{FP}) \approx  N \gamma_{FP} / \gamma=1$ which simply gives:
\begin{equation}
\gamma_{FP}=T_0\gamma
\end{equation}

We took the case of a highly reflecting mirror for which $T_0$ is very small and thus the radiative linewidth at fabry perot resonance($\gamma_{FP}$) is much less than the radiative linewidth at excitonic resonance ($\gamma$). Due to the small linewidth we have a much longer lifetime of the polariton. The long lifetime is a result of the fact that exitonic radiation and the photonic component of the polariton destructively interfere.

The plot below gives the max transmission cooficient ($T_{MAX}$) as a function of the separation $d$. For this we sweep the detuning for each value of $d$ and note the maximum value for that $d$. We see that for values of d close to integer multiples of $\lambda/2$ we observe perfect transmission. These values are the fabry perot distance $d_{FP}$

```python
max_pt=np.linspace(0,0,400)
d_points=np.linspace(0,1,400)
for i in range(400):
    y=[f(j,d_points[i]) for j in x]
    max_pt[i]= max(y)
plt.plot(d_points,max_pt) 
plt.ylabel("$T_{MAX}$")
plt.xlabel(r'$d/\lambda$')
plt.show() 
#plot by Abhishek
```

Polariton's are primarily composed of excitonic degrees of freedom since we are considering a weak field regime. Thus, inter-polariton interactions can be approximated by the inter-exciton interactions in the absence of the mirror.

But, an important property to be noted is that the lifetime of the polariton is much more than that of the exciton due to reasons explained above.

In the ansence of the mirror, we were not able to see the non-linear effects since that required to have a non linear interaction parameter $\chi$ $>$ $\gamma$ (the resonance linewidth). In the presence of the mirror the resonance linewidth becomes $\gamma_{FP}$ $=$ $T_0\gamma$ (linewidth for fabry perot resonance). Thus to observe non linear effects in the presence of a mirror we require $\chi$ $>$ $T_0\gamma$ which in turn causes $1/T_0$ fold enhancement of non-linearity. We will see that this non linearity leads to photon antibunching thus making the system act like a non classical light source (single photon source).

To get the governing equation of the light matter interaction due to the presence of the mirror we start by writing the net hamiltonian of the scenario which includes the system hamiltonian, the bath hamiltonian (describing the dynamics of the photon modes), and the interaction hamiltonian (which involves coupling of photons to the excitonic modes). This hamiltonian is given by :

\begin{equation}
H=H_{0}+\sum_{k} \omega_{k} d_{k}^{\dagger} d_{k}+i \sum_{k} g_{k}\left(d_{k}^{\dagger} a-a^{\dagger} d_{k}\right)
\end{equation}
where $d^{\dagger}_k$ ($d_k$) are the photon creation (annihilation) operators.
Till now we have only taken the scenario of free space photon modes. The mirror can be introduced into the system by taking appropriate boundary conditions. 

Now, by using the heisenberg's equations of motion (taking commutator bracket of operators with the above hamiltonian) we can derive the dynamics of the photon operators ($d_k$) and the system operators ($a$).
Solving these equations and including the appropriate boundary conditions gives us the equation defining the dynamics of an arbitrary system operator $\textbf{Q}$. This equation is known as the **Heisenberg-Langevin** equation:

\begin{equation}
\begin{aligned}
\dot{Q}=&-i\left[Q, H_{0}+\frac{\gamma}{2} \operatorname{Im}\left(r_{0} e^{2 i k_{0} d}\right) a^{\dagger} a+\Omega a^{\dagger}+\Omega^{*} a\right] \\
&+\mathcal{D}[Q]+\mathcal{F}[Q] \\
\text { where } & k_{0}=\omega_{0} / c \text { and } \\
\end{aligned}
\end{equation}

The term $\Omega$ which is known as the Rabi Frequency is given by:

\begin{equation}
\begin{aligned}
\Omega=& \sqrt{\frac{\gamma}{2}}\left(1+r_{0} e^{2 i k_{0} d}\right)\left\langle b_{\text {in }, R}\right\rangle+\sqrt{\frac{\gamma}{2}} t_{0} e^{i k_{0} d}\left\langle b_{\text {in }, L}\right\rangle
\end{aligned}
\end{equation}
This is a superposition of the two input fields $b_{in,R}$ which is input towards right and $b_{in,L}$ which is input towards the left. These fields freely propagate as if there was no TMD and mirror present. See figure(1) for a better understanding.

The term $D[Q]$ describes the dissipations in the system. It is given by:
\begin{equation}
\begin{aligned}
\mathcal{D}[Q]=&\left[\gamma+\gamma \operatorname{Re}\left(r_{0} e^{2 i k_{0} d}\right)+\gamma^{\prime}\right]\left(a^{\dagger} Q a-\frac{1}{2}\left\{Q, a^{\dagger} a\right\}\right) \\
&+\frac{\chi_{2}}{2}\left(a^{\dagger} a^{\dagger} Q a a-\frac{1}{2}\left\{Q, a^{\dagger} a^{\dagger} a a\right\}\right)
\end{aligned}
\end{equation}
The term $F[Q]$ is known as the langevin noise term which has an interesting property that it's expectation value $\left\langle F[Q]\right\rangle$ is $0$ regardless of the choice of $Q$

We have encountered the linear terms $\gamma$ and $\gamma^{\prime}$ previously but here we see an additional dissipation term $\chi_{2}$ which gives the non-radiative decay due to exciton-exciton interactions.

**Master Equations in quantum optical systems**

We have a system S which is weakly coupled to a bath B. The Hamiltonian of the combined system is given by Hs + Hb + Hint. Performing a unitary transformation $U(t)=\exp \left[-i\left(H_{S}+H_{B}\right) t\right]$
The typical timescale of the dynamics generated by this master equation is characterized by $\tau_R$, which is the relaxation time of the system due to the interaction with the bath. For the Markov approximation to be valid, neglecting any memory effects in the system, which can be characteristic timescale ${\displaystyle \tau _{B}}$ under the approximation that correlations in the bath decay within $\tau_B$.this relaxation time has to be long compared to the bath correlation time, i.e, ${\displaystyle \tau _{R}\gg \tau _{B}}$. For quantum optical systems, ${\displaystyle \tau _{B}}$ is the inverse of the optical frequency, i.e., several inverse THz, while the lifetime of an optical excitation is in the inverse MHz range. Therefore, the Markov approximation is well justified in quantum optical systems.


**Optical cavity: quantum Langevin equations**

Model for the coupling of a quantized cavity mode to the continuum of modes outside. To this end we consider the Hamiltonian for the cavity mode plus the infinity of outside modes and their interaction (modelled a an excitation exchange term)
$$
H=\hbar \omega_{c} a^{\dagger} a+\sum_{k} \hbar \omega_{k} b_{k}^{\dagger} b_{k}+\hbar \sum_{k} g_{k}\left[b_{k}^{\dagger} a+a^{\dagger} b_{k}\right]
$$
From here we can derive a set of equations of motion Let's write the equations of motion:
$$
\begin{aligned}
\dot{a}=\frac{i}{\hbar}[H, a] &=-i \omega_{c} a-i \sum_{k} g_{k} b_{k} \\
\dot{b}_{k} &=\frac{i}{\hbar}\left[H, b_{k}\right]=-i \omega_{k} b_{k}-i g_{k} a
\end{aligned}
$$
The plan is to formally integrate the equation of motion for the outside modes and plug the solutions back into the cavity mode evolution equation. This will result in a differential-integro equation. In a first step we have
$$
b_{k}(t)=b_{k}(0) e^{-i \omega_{k} t}-i g_{k} \int_{0}^{t} d t^{\prime} a\left(t^{\prime}\right) e^{-i \omega_{k}\left(t-t^{\prime}\right)}
$$
Replacing this expression into the cavity mode equation of motion we obtain:
$$
\dot{a}=-i \omega_{c} a-\sum_{k} g_{k}^{2} \int d t^{\prime} a\left(t^{\prime}\right) e^{-i \omega_{k}\left(t-t^{\prime}\right)}-i \sum_{k} g_{k} b_{k}(0) e^{-i \omega_{k} t}
$$
The second term is an infinite sum over the input modes which we will denote by
$$
F(t)=-i \sum_{k} g_{k} b_{k}(0) e^{i-\omega_{k} t}
$$ $F(t)$ is the Langevin Noise operator of the open quantum system.

The term ‘Anti-Bunching’ refers to light the photons of which appear at more regular times than in a classic light signal. In a light signal emitted by a classic source, such as thermal emission from a black body or fluorescence from a large number of molecules, the photons are emitted at random times. After excitation, a single emitter cannot be re-excited (and give off a second photon) until it has returned from the excited state. ###Derivation of the Heisenberg Langevin equations

For a system interacting with a bath (here a source of photons), the net hamiltonian is given by:
\begin{equation}
H=H_{0}+\sum_{k} \omega_{k} d_{k}^{\dagger} d_{k}+i \sum_{k} g_{k}\left(d_{k}^{\dagger} a-a^{\dagger} d_{k}\right)
\end{equation}
where $d^{\dagger}_k$ ($d_k$) are the photon creation (annihilation) operators. $H_0$ is the system hamiltonian and for the scenario involving non linear interactions. It is given by:
\begin{equation}
H_{0}=\omega_{0} a^{\dagger} a+\left(\chi_{1} / 2\right) a^{\dagger} a^{\dagger} a a
\end{equation}
where $a^{\dagger}(a)$ are the exciton creation (annihilation) operators.

We first derive this for free space and then introduce the mirror by applying appropriate boundary conditions.

In a more general context we can write the hamiltonians in the form of intergrals over frequencies the governing hamiltonian becomes:
\begin{equation}
\begin{array}{l}
H=H_{\mathrm{0}}+H_{B}+H_{\mathrm{int}} \\
where,\\
H_{B}=\hbar \int_{-\infty}^{\infty} d \omega \omega b^{\dagger}(\omega) b(\omega) \\
H_{\mathrm{int}}=i \hbar \int_{-\infty}^{\infty} d \omega \kappa(\omega)\left[b^{\dagger}(\omega) c-c^{\dagger} b(\omega)\right]
\end{array}
\end{equation}
$\kappa(\omega)$ is the coupling between the system and the photons. here $c^{\dagger}$ is the system operator and $b^{\dagger}$ are the photon creation operators.

Now, in the heisenberg picture, the dynamics of the operators is given by the commutator bracket of the operator with the governing hamiltonian. 
\begin{equation}
\begin{array}{l}
\dot{b}(\omega)=-i \omega k(\omega)+\kappa(\omega) c \\
\dot{a}=-\frac{i}{\hbar}\left[a, H_{\text {0 }}\right]+\int d \omega \kappa(\omega)\left\{b^{\dagger}(\omega)[a, c]\right.\left.\quad-\left[a, c^{\dagger}\right] b(\omega)\right\}
\end{array}
\end{equation}
Here $a$ is any arbitrary system operator.

Solving for $\dot{b}$ gives:
\begin{equation}
b(\omega)=e^{-i \omega\left(t-t_{0}\right)} b_{0}(\omega)+\kappa(\omega) \int_{t_{0}}^{t} e^{-i \omega\left(t-t^{\prime}\right)} c\left(t^{\prime}\right) d t^{\prime}
\end{equation}

We can put this integral to the relation for $\dot{a}$. On doing this we get:
\begin{equation}
\begin{array}{r}
\dot{a}=-\frac{i}{\hbar}\left[a, H_{\mathrm{0}}\right]+\int d \omega \kappa(\omega)\left\{e^{i \omega\left(t-t_{0}\right)} b^{\dagger}(\omega)[a, c]\right. +\int d \omega[\kappa(\omega)]^{2} \int_{t_{0}}^{t} d t^{\prime}\left\{e^{i \omega\left(t-t^{\prime}\right)} c^{\dagger}\left(t^{\prime}\right)[a, c]\right. \left.-\left[a, c^{\dagger}\right] e^{-i \omega\left(t-t^{\prime}\right)} c\left(t^{\prime}\right)\right\}
\end{array}
\end{equation}

Now we make an important approximation known as the Markov's approximation which states that the interaction cooficient $\kappa(\omega)$ is independent of the frequency and is given by : $\kappa(\omega)=\sqrt{\gamma/2\pi}$ . This simplifies the equation significantly.

We introduce a term in field which is given by:
\begin{equation}
b_{\text {in }}(t)=\frac{1}{\sqrt{2 \pi}} \int d \omega e^{-i \omega\left(t-t_{0}\right)} b_{0}(\omega)
\end{equation}

Using the relations defined above we finally get:
\begin{equation}
\begin{aligned}
\dot{a}=-\frac{1}{\hbar}\left[a, H_{\mathrm{0}}\right]-&\left[\left[a, c^{\dagger}\right]\left[\frac{\gamma}{2} c+\sqrt{\gamma} b_{\mathrm{in}}(t)\right]\right]- \left[\frac{\gamma}{2} c^{\dagger}+\sqrt{\gamma} b_{\mathrm{in}}^{\dagger}(t)\right][a, c]
\end{aligned}
\end{equation}

The above equation is the quantum langevin equation.

Coming back to our system we apply the same procedure to get the evolution equation of any arbitrary system operator $Q$. The equation given below doesn't account for the presence of the mirror which will be taken care of by taking appropriate boundary conditions.

By the procedure above we get the equation to be:
\begin{equation}
\begin{aligned}
\dot{Q}=-i\left[Q, H_{0}\right] &+\gamma\left(a^{\dagger} Q a-\frac{1}{2}\left\{Q, a^{\dagger} a\right\}\right) \\
&+\sqrt{\frac{\gamma}{2}}\left(b_{\mathrm{in}, R}^{\dagger}[Q, a]-\left[Q, a^{\dagger}\right] b_{\mathrm{in}, R}\right)+\sqrt{\frac{\gamma}{2}}\left(c_{\mathrm{in}}^{\dagger}[Q, a]-\left[Q, a^{\dagger}\right] c_{\mathrm{in}}\right)
\end{aligned}
\end{equation}

where $b_{in,R}$ and $c_{in}$ are defined by :
\begin{equation}
b_{\mathrm{in}, R}(t)=\sum_{k>0} d_{k}\left(t_{0}\right) e^{-i \omega_{k}\left(t-t_{0}\right)}, \quad c_{\mathrm{in}}(t)=\sum_{k<0} d_{k}\left(t_{0}\right) e^{-i \omega_{k}\left(t-t_{0}\right)}
\end{equation}

The equations given below describe the boundary conditions in the presence of a mirror.
For left moving modes:
\begin{equation}
E_{R}^{+}(z, t) \propto \sum_{k>0} g_{k} d_{k}(t) e^{i k z} \propto\left\{\begin{array}{ll}
b_{\text {in, } R}(t-z / c) & \text { if } z<0 \\
c_{\text {out }}(t-z / c) & \text { if } 0<z<d \\
b_{\text {out }, R}(t-(z-d) / c) & \text { if } d<z
\end{array}\right.
\end{equation}
and for right moving modes:
\begin{equation}
E_{L}^{+}(z, t) \propto \sum_{k<0} g_{k} d_{k}(t) e^{i k z} \propto\left\{\begin{array}{ll}
b_{\text {out }, L}(t+z / c) & \text { if } z<0 \\
c_{\text {in }}(t+z / c) & \text { if } 0<z<d \\
b_{\text {in }, L}(t+(z-d) / c) & \text { if } d<z
\end{array}\right.
\end{equation}
where \begin{equation}
c_{\text {out }}(t)=b_{\text {in }, R}(t)+\sqrt{\frac{\gamma}{2}} a(t), \quad b_{\text {out }, L}(t)=c_{\text {in }}(t)+\sqrt{\frac{\gamma}{2}} a(t)
\end{equation}

Taking all these terms into account gives us the final equation governing our system:
\begin{equation}
\dot{Q}=-i\left[Q, H_{0}+\frac{\gamma}{2} \operatorname{Im}\left(r_{0} e^{2 i k_{0} d}\right) a^{\dagger} a-i\left(\Omega a^{\dagger}-\Omega^{*} a\right)\right]+\gamma\left[1+\operatorname{Re}\left(r_{0} e^{2 i k_{0} d}\right)\right] a^{\dagger} Q a-\frac{1}{2}\left[Q, a^{\dagger} a\right]+\mathcal{F}_{Q}
\end{equation}
where $\Omega$(Rabi Frequency) is given by: $\Omega=\sqrt{\frac{\gamma}{2}}\left(1+r_{0} e^{2 i k_{0} d}\right)\left\langle b_{\mathrm{in}, R}\right\rangle+\sqrt{\frac{\gamma}{2}} t_{0} e^{i k_{0} d}\left\langle b_{\mathrm{in}, L}\right\rangle$
