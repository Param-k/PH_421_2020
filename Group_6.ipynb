##<center><font color=red> $(1)$ **Introduction and Background** </font></center>

Non-linear optical materials have a variety of practical applications, and the topic of nonlinear optics has been covered in quite some depth in the present course. However, it turns out that the susceptibility tensor we encountered in photonics is pretty complicated and it's often not possible to make observations beyond basic inferences regarding the symmetry properties. The topological formulation in terms of the Berry connection and Berry curvature allows us to robustly connect the nonlinear optical responses of the material to the microscopic interactions by framing the properties in terms of topological invariants. 

Topological invariants do not change under slight deformations as those properties depend upon integers, and small deformations can only affect parameters which can be continuously varied. This makes it useful to undertake a study of the topological nature of the above-mentioned nonlinear effects.

This has applications in recent discoveries for solar cells with perovskite oxides with non-centrosymmetric crystal structure which have been shown to have large efficiencies. We closely follow the analysis in the paper assigned to us $^{[40]}$.  The current response has been captured in several classical papers using the Keldysh formulation which is also used in this paper. The bulk photovoltaic effect (described later) shown by this crystal is caused by a "shift current", leading to a DC flow of electrons along the crystal. As electrons translate in the unit cells, their "intracell" coordinates (intuitively describing their positions within a unit cell) change. The shift vector (derived in the present report) is a measure of this quantity, and is given in terms of Berry connections.


The model that we use in this paper consists of a crystal structure. We focus on processes that involve 2 bands such as the shift current. The crystal interacts with monochromatic light which takes the form of a photon with frequency $\Omega$. The system is further coupled to a fermionic heat bath at every site to ensure that we reach a non equilibrium steady state, The fermions in the system are assumed to be non interacting, so we ignore any repulsion between the electrons in the current.  



###<center>
1.1 Green's Functions in Non-Equilibrium Quantum Mechanics</center>

The Green's function (often called the propagator) in standard quantum mechanics is the matrix element of the unitary time-evolution operator $U(t',t)$, in the position basis: $G(x',x)=\langle x'|U(t',t)|x\rangle$. As can be easily observed, $\int \text{d}x G(x',x)\psi(x,t)=\int \text{d}x\, \langle x'|U(t',t)|x\rangle\langle x|\psi (t)\rangle=\langle x'|U(t',t)|\psi(t)\rangle=\psi(x',t')$, establishing an analogy with the definition oft-encounted in maths.

Our focus in this section is to introduce the Keldysh Green's function as a means to attack problems in non-equilibrium many-body quantum mechanics. We do this via the path integral route. We shall see that an elegant correspondence will emerge between the lesser Green's function, $G^<$, and the density matrix of the system in question.

We start off defining the four Green's functions we'll employ in this formalism. We handle time dependence in the Heisenberg picture, so that operators evolve as $\mathcal O(t)=U^\dagger(t) \mathcal O U(t)$.


The general scheme to deal with many body problems in QM $^[5]$is to assume that we have a non-interacting Hamiltonian $\hat{H}_0$, an interacting hamiltonian $H_1$ and an external Hamiltonian $\hat{H}_{ext}$. It is assumed that at time  $t = -\infty$ the system is in the noninteracting ground state $|\rangle_0$. The interactions are then adiabatically switched on so that at time $t = 0$ the system comes to the ground state of $|0\rangle$ of the hamiltonian $\hat{H} = \hat{H}_0 + \hat{H}_1$.This evolution is given definitionally by the $S$-matrix operator. The S matrix holds the scattering data for the process that connects the in and out states based on the interactions in the hamiltonian. 
$$
\hat{S}(t,t') = e^{i\hat{H}_0t}e^{-i\hat{H}(t-t')}e^{-i\hat{H}_0t'} 
$$
whose action turns the interactions on at time t'. The central term represents the time evolution unitary in the Schrodinger picture; the unitaries multipying it from the left and right, help us shift the unitary to the interaction picture. The effect on the noninteracting ground state is given by
$$
|0\rangle = \hat{S}(0,-\infty)|\rangle_0
$$
When the system is in equilibrium,  after swtiching on the interactions and then adiabatically them swtiching off at some later time, the final state reached at $t = +\infty$ is going to be  $|\infty\rangle$, i.e. 
$$
|\infty\rangle = \hat{S}(+\infty,-\infty)|\rangle_0 
$$
This allows us to to write down the expression for the correlators$^{[1][5]}$ of operators given by 
$$
C(t,t') = \phantom{1}_0\langle|\mathcal T\hat{S}(-\infty,0)\hat{S}(0,t)\hat{A}(t)\hat{S}(t,t')\hat{B}(t')\hat{S}(t',0)\hat{S}(0,-\infty)|\rangle_0
$$
This makes sense from the textbook QFT point of view, where correlators are given by $\langle 0|\mathcal T \phi_1\phi_2|0\rangle$. Here, $\mathcal T$ is the time ordering operator. What happens here is,

i)The states $|\rangle_0, \phantom{1}_0\langle |$ are evolved by the interaction hamiltonian to time $t,t'$ respectively.

ii)The operators $A,B$ act on the states at time $t',t$ respectively. Finally the state is evolved from $t$ to $t'$

An example of this would be where $\hat{A},\hat{B}$ correspond to the annhilation and creation operators. 

The correlator can be written more succintly as 
$$
C(t,t') = \phantom{1}_0\langle|\mathcal T\hat{S}(+\infty,-\infty)\hat{A}(t)\hat{B}(t')|\rangle_0
$$
The time ordering operator rearranges these operators in increasing order of time to get back the orignal expression.


The final state is the same in equilibrium, irrespective of how the interactions are swtiched on and off. This is no longer true when the system is out of equilibrium, which creates problems when we want to evaluate the correlators, as it is difficult to evaluate where the state could be, at time $t = +\infty$ when the interactions are switched off at some time in future. 

In order to avoid the difficulty of finding the state after interactions are switched off at  $t = +\infty$ we use the idea of a Keldysh Contour$^{[2][3][5]}$: We first evolve the state from $t = -\infty$ to $+\infty$ and then unwind the time evolution by going back to $t = -\infty$. Therefore, the switiching on and switiching off of the interactions both happens in the past. 

This however means that we must define the field operators on 2 branches, on $-\infty \to +\infty$ and on $+\infty \to -\infty$, doubling the degrees of freedom.

The above figure shows the keldysh contour.

Label the field $\phi_+(t)$ if it belongs to the forward contour and $\phi_-(t)$ if it belongs to the backwards contour. 



On the Keldysh contour we can define the partition function in terms of the action of the system $S$
$$
Z = \int \mathcal{D}\phi \bar{\phi} \phantom{1}e^{iS[\phi,\bar{\phi}] }
$$
In order to understand why we obtain a functional integral, look at appendix A. Note that as defined previously, $Z = \frac{\mathbf{tr}(\hat{U}_\mathcal{C}\rho_0)}{\mathbf{tr}(\rho_0)}$. We can evaluate this trace by first breaking the contour into discrete sections and then taking a continuum limit. The continuum limit gives us the Feynman path integral, denoted by $Z$ due to its correspondence with the partition function from statistical mechanics.  
The Green's functions take the form 
\begin{equation}
    G(t,t') = -i \langle \phi(t)\bar{\phi}(t') \rangle 
\end{equation}
Which correspond to the field operators. Explicitly, $\phi,\bar{\phi}$ correspond to the eigenvalues of the annhilation and creation operators $a|\phi\rangle = \phi |\phi\rangle,\langle\phi|a^\dagger = \langle\phi|\bar{\phi}$. The exact expression for this Green's function takes the form 
$$
G(t,t') = \int \mathcal{D}\phi \bar{\phi} e^{iS[\phi,\bar{\phi}] }\phi(t)\bar{\phi}(t)
$$
Notice that we can define 4 different types of Green's functions on the Keldysh contour based on where the field operators are on the keldysh contour. In the density matrix picture, recall that observables are given by traces :  $\langle \hat{O}\rangle = \frac{\mathbf{tr}(\hat{O}\rho)}{\mathbf{tr}(\rho)}$. We can therefore define the Green's functions as: 
$$
1)\phantom{1} iG^{<}(t,t') = \langle \phi_+(t)\bar{\phi}_-(t')\rangle = \frac{\mathbf{tr}(a^\dagger(t')a(t)\hat{\rho_0})}{\mathbf{tr}(\hat{\rho}_0)}
$$
Specifically, because the fields are on different contours, we can ignore the time ordering as anything on the negative contour comes after the positive contour. Similarly
$$
2)\phantom{1} iG^{>}(t,t') =  \langle \phi_-(t)\bar{\phi}_+(t')\rangle
$$
$$
3)\phantom{1} iG^{T}(t,t') =  \langle \phi_+(t)\bar{\phi}_+(t')\rangle = \frac{\mathbf{tr}(Ta^\dagger(t')a(t)\hat{\rho_0})}{\mathbf{tr}(\hat{\rho}_0)}
$$

When both fields are on the forward branch, we must use the time ordering operator. 
$$
4)\phantom{1} iG^{\tilde{T}}(t,t') =  \langle \phi_-(t)\bar{\phi}_-(t')\rangle = \frac{\mathbf{tr}(\tilde{T}a^\dagger(t')a(t)\hat{\rho_0})}{\mathbf{tr}(\hat{\rho}_0)}
$$
When both fields are on the backward branch, we are going from $+\infty \to -\infty $ So we need to use the anti-time ordering operator $\tilde{T}$

The 4 green's functions defined here are not independent 

$$
G^{T} + G^{\tilde{T}} = G^{<} + G^{>}
$$

This fact is used to do a change of variables called the Keldysh rotation. The form of the Keldysh rotation for bosons is given by 

$$
\phi_1 = \frac{1}{\sqrt{2}}\left( \phi_+(t) + \phi_-(t)  \right)
$$

$$
\phi_2 = \frac{1}{\sqrt{2}}\left( \phi_+(t) - \phi_-(t)  \right)
$$

And the keldysh rotation for fermions is given by 

$$
\psi_1(t) = \frac{1}{\sqrt{2}} \left(\psi_{+}(t) + \psi_{-}(t) \right), \psi_2(t) = \frac{1}{\sqrt{2}} \left( \psi_{+}(t) - \psi_{-}(t) \right)
$$

However, the transformation for the creation opertor fields given by $\langle\psi|c^{\dagger}= \langle\psi|\bar{\psi}$ is given by 

$$
\bar{\psi}_1(t) = \frac{1}{\sqrt{2}} \left(\bar{\psi}_{+}(t) - \bar{\psi}_{-}(t) \right), \bar{\psi}_2(t) = \frac{1}{\sqrt{2}} \left(\bar{\psi}_{+}(t) + \bar{\psi}_{-}(t) \right), 
$$
WARNING: THE PROPERTIES AND INTERPRETATIONS OF $\phi_1,\phi_2$ VARY BASED ON WHETHER THE FIELDS ARE BOSONIC OR FERMIONIC. For the rest of the paper we are interested in the fermion picture but there are a lot of interesting applications for bosons also. 

 As a consequence of the Keldysh rotation the Green's functions are recast in the following for fermions:  

$$
G^{\alpha\beta}(t,t') = \begin{pmatrix}
    G^{R}(t,t') && G^{K}(t,t') \\
    0 && G^{A}(t,t')  
\end{pmatrix}
$$
Where $\alpha,\beta$ may be $1$ or $2$ and the rotated Green's functions are given by 
$$
G^R = \frac{1}{2} (G^T - G^{\tilde{T}}-G^< + G^> ) = \theta(t-t')(G^> \pm G^<)
$$

$$
G^A = \frac{1}{2} (G^T - G^{\tilde{T}}+G^< - G^> ) = \theta(t'-t)(G^< \pm G^>)
$$

$$
G^K = G^< + G^>
$$

Where $A,R,K$ stand for the Advanced, Retarded and Keldysh components of the Green's functions respectively. 

##Dyson Equation 
The power of this method becomes clear when we add in interactions $^{[3][4]}$ to our non-interacting action. 
$$
S = S_o + S_{int}
$$
The Green's function for this action are given by 
$$
\mathbf{G}^{\alpha\beta}(t,t') = \int \mathcal{D}\phi \bar{\phi} e^{iS_0+S_{int} }\phi(t)\bar{\phi}(t)
$$
The full Green's function in the interaction picture can be given by the expansion of the path ordered exponential, which gives us 
$$
\mathbf{G}(t,t') = \frac{\sum_{n = 0}^{\infty} \frac{1}{n!}(-i)^n \int dt_1 \int dt_2... \int dt_n T\left[\hat{H}_{int}(t_1) \hat{H}_{int}(t_2).. \hat{H}_{int}(t_n)\psi(t)\bar{\psi}(t')  \right]}{\sum_{n = 0}^{\infty} \frac{1}{n!}(-i)^n \int dt_1 \int dt_2... \int dt_n T\left[ \hat{H}_{int}(t_1) \hat{H}_{int}(t_2).. \hat{H}_{int}(t_n)  \right]}
$$

The time ordered product can be simplified using Wicks theorem $^{[1]}$ which states that the time ordered product of $n$ given fields equals the normal ordered product, plus all possible normal ordered products which result from `contracting' pairs of fields in the product. Contraction refers removing the fields from the product, and multiplying the product with their correlator. 
Treating the interactions as a perturbation, we can express the green's functions in terms of Feynman Diagrams.

\int_{-\infty}^{+\infty}dt_1[G_0^{-1}(t,t_1) - \Sigma(t,t_1)].G(t_1,t') = \mathbb{I}_{2\times 2}\delta(t,t')

$$
\mathbf{G} = G + G\circ\Sigma\circ G + G\circ\Sigma\circ G\circ\Sigma\circ G +...
$$
Where $\Sigma$ corresponds to the irreducible self energy, ie the sum of those Feynman diagrams which cannot be separated into 2 diagrams by cutting a single solid line (IE cannot be decomposed into Energy - Propagator - Energy). Keldysh in Flouqet formalism 

When the unperturbed hamiltonian is periodic in time, there is additional redundancy in the Green's functions. $^{[6][7][8]}$
$$ 
G(t_r,t_{avg}) = G(t_r,t_{avg}+\tau)
$$ 
where $t_r= t_1-t_2$ , $t_{avg} = \frac{t_1+t_2}{2}$ and $\tau $is the periodicty. (Essentially this means that we are transposing $t_1, t_2$ to $t_1+\tau, t_2+\tau$). This degeneracy can be avoided by going to the floquet representation of a function: 
$$
F_{mn}(\omega) = \frac{1}{\tau}\int_0^{\tau}dt_{avg} \int_{-\infty}^{+\infty} dt_re^{(i\omega+\frac{m+n}{2}\Omega)t_r}e^{i(m-n)\Omega t_{avg}}F(t_r,t_{avg})  
$$
Where $\Omega = \frac{2\pi}{\tau}$ . In this representation the $\circ$ in the dyson reduces to a matrix multiplication at each $\omega$. 
This is related to the Wigner representation which is given by 
$$
F_n(\omega) = \frac{1}{\tau}\int_{0}^{\tau}dt_{avg} \int_{-\infty}^{+\infty}dt_r e^{i\omega t_r}e^{in\Omega t_{avg}}F(t_r;t_{avg}) 
$$
Clearly
$$
F_{mn} (\omega) = F_{m-n}(\omega + \frac{m+n}{2}\Omega)
$$

The dyson equation now reduces to 
$$
\mathbf{G}^{-1}(\omega)_{mn} = G_0^{-1}(\omega)_{mn} + \Sigma_{mn}  
$$
This is because the floquet transformation reduces the convolution product to a matrix product which is proved below $^{[8]}$
$$
\int_{-\infty}^{+\infty}dt'' A(t,t'')B(t'',t') = \int dt'' \sum_l \int\frac{d\omega}{2\pi}e^{-i\omega(t-t'')-il\Omega(t+t'')/2}A_l(\omega) \sum_{l'}\int \frac{d\omega '}{2\pi}e^{-\omega '(t''-t') - il'\Omega(t'+t'')/2}B_{l'}(w')
$$
This expression is obtained by doing an inverse Wigner transform. We can reduce this integral by exchanging the order of integration. 
$$
\int_{-\infty}^{+\infty}dt'' A(t,t'')B(t'',t') = \sum_{l,l'} \int \frac{d\omega}{2\pi}\int \frac{d\omega '}{2\pi} \int dt'' e^{-i\omega(t-t'')-il\Omega(t+t'')/2}e^{-\omega '(t''-t') - il'\Omega(t'+t'')/2}A_l(\omega)B_{l'}(w') 
$$
The only part dependent on $t''$ is the exponential. The integral wrt $t''$ therefore gives us $2\pi\delta(\omega - \omega ' - \frac{l+l'}{2}\Omega )$ (Which is a very standard result in fourier theory). 
The integral wrt $\omega'$ therefore picks up only the value of the function when the delta function above is 0 ie $\omega' = \omega -\frac{l+l'}{2}\Omega$. 
The final expression therefore is 

$$
\int_{-\infty}^{+\infty}dt" A(t,t")B(t",t') = \sum_{l,l'} \int \frac{d\omega}{2\pi}  e^{-i(\omega - l'\Omega/2)(t-t') - i(l+l')\Omega(t+t')/2}A_l(\omega)B_{l'}(\omega - -\frac{l+l'}{2}\Omega) 
$$
On comparing this with the wigner transform for the function $(A*B)(t,t')$ we get 
$$
(AB)_k(\omega) = \sum_{l+l' = k}(\omega) = \sum_{l+l' = k} A_l(\omega + \frac{l'}{2}\Omega)B_{l'}(w-\frac{l}{2}\Omega)
$$
We can then make a transformation $w \to w + \frac{k'}{2}\Omega $ Where $k',k$ have the same remainder wrt 2(This is to ensure $(k+k')/2$ and  $(k-k')/2$ are both integers). This allows us to write down the equation in the floquet representation giving us 
$$
(AB)_{mn} (\omega) = \sum_{m'} A_{mm'}\omega B_{m'n}(\omega)
$$  
Where $m= \frac{k+k'}{2} $, $ n = \frac{k-k'}{2}$, $ m' = \frac{k+k'}{2}-l$. This means that the convolution has reduced to a matrix multiplation over floquet indices.

Derivation Of Self Energy for Fermi Heat bath 

The hamiltonian for the heat bath is assumed to be a fermi gas which satisfies Fermi-Dirac statistics: 
$$
H_{bath} =  \sum (\epsilon_{a,p} - \mu_a)a_{i,p}^{\dagger}a_{i,p}
$$
Here the operators $a^\dagger ,a $ create and annhilate fermions in the heat bath,$\epsilon_{a,p}$ is the energy associated with the $p^{th}$ excited state at the $i^{th}$ site. $\mu_a$ is the chemical potential. The statitics followed are the usual fermi dirac statistics.  
$$
\langle \Psi_0 |a^\dagger_{k} a_{k} |\Psi_0\rangle = f_{k} = \begin{cases}
    1 & \epsilon_k < \epsilon_F \\
    0 & \epsilon_k > \epsilon_F \\
\end{cases}  
$$
Where $|\Psi_0\rangle$ is the ground state of the fermionic bath. 
The coupling of the heat bath with the system is represented by the following hamiltonian 
$$
H_{int }= \sum_{i,p}V_p c_i^{\dagger}a_{i,p} +$$ Hermitian Conjugate

Where $c_i^\dagger$ creates a fermion in the system. The index i represents the fact that the heat bath is coupled to every site in the system. $V_p$ is a coupling constant that is related to the thermal dissipation in our system with $|V_p|^2 ~ \Gamma $. 

The green's function within the bath are easy to calculate as the electrons are assumed to be non interacting, Thus the usual green's function of a non-interacting fermion. 

The self energy correction then becomes 
$$
\Sigma(t,t') = \sum_p V_p G_{bath}(p;t,t') V_p 
$$
As the particles are non interacting, we can therefore directly substitute the $G^R$ for the free fermion to In the frequency representation, we get 
$$
\Sigma_{bath}^{R}(w) = \sum_{p} = \frac{V_p^2}{\omega + \mu_b -\epsilon_{b,p} +i\eta}   
$$
where $\eta$ is an infitesimal parameter we add to ensure the fourier transformation converges. This parameter can be removed by breaking this fraction down into it's real and imaginary part as follows 
$$
\frac{1}{\omega + i\eta} = \mathcal{P}(\frac{1}{\omega}) - i\pi \delta(\omega)
$$
Where $\mathcal{P}$ denotes the principal part. The imaginary part of this function therefore looks like 
$$
\Gamma(\omega) = \sum_p \pi V_p^2 \delta (\omega + \mu_b - \epsilon_{b,p}) 
$$
. Only the imaginary part matters as the real part can be absorebed into the chemical potential of the system to ensure that no net current passes through the interface.  The fact that the dynamics depend on the imaginary part of the self energy indicates that the interaction with the heat bath is dissipative and irreversible. 

$\Gamma$ is a spectral function on the heat bath: Notice that $\Gamma$ seems to be nonzero for very specific frequencies corresponding to the excitions $\omega = \epsilon - \mu $. As the heat bath can be assumed to be in equilibrium, we can use the relation between the green's function and the spectral function to derive the full Self energy matrix. As the self energy is proportional the green's functions, we can use the fluctuation dissipation relation.

G^< = G^R \Sigma ^< G^A

And

###<center>1.2 The Floquet Formalism</center>

####<center>Motivation for using Floquet systems

The time-dependent systems we study are often periodic-- such as electromagnetic field oscillating at a particular frequency. While it is possible to analyse their dynamics by numerically integrating the Schrodinger's equation, it is much more elegant to convert them into time-independent problems and then analyse them. One method of doing so is the Floquet formalism.

####<center>The Theory of Differential Equations

Floquet theory deals with the solutions to periodic linear ordinary differential equations of the form:

$$ \dot{x} = A(t)x $$
where A(t) is a piecewise continuous periodic function with period 
T and we assume all functions are well-behaved.

Floquet Theory, developed by Gaston Floquet (1883), gives a canonical form for each fundamental matrix solution of this common linear system. When applied to physical systems with periodic potentials, such as crystals in condensed matter physics, the result is known as Bloch's theorem.

####<center>Floquet theorem for a time-dependent Periodic Hamiltonian

We apply this theory to our system with a time-dependent periodic Hamiltonian $H(t)$, such that $H(t)=H(t+T)$. The Schrodinger equation is:
$$
H(t) \Psi(t)=i \hbar \frac{\partial}{\partial t} \Psi(t).
$$
According to the Floquet theorem, the solutions are of the form
$$
\Psi_{\alpha}(t)=\exp \left(-i \epsilon_{\alpha} (t-t_0) / \hbar\right) \Phi_{\alpha}(t).
$$
Here, $\Psi_{\alpha}(t)$ represents the Floquet states which are the solutions to the Schrödinger equation.
$\Phi_{\alpha}(t)=\Phi_{\alpha}(t+T)$ are the periodic Floquet modes, which provide a complete and orthonormal basis for the Hilbert space. 
$\epsilon_{\alpha}$ are the quasienergy levels. The quasienergy levels are constants in time, but only uniquely defined up to multiples of $ 2 \pi\hbar / T$ . ${[23]}$

If we know the Floquet modes (for $t \in[0, T]$ ) and the quasienergies for a particular $H(t)$, we can easily decompose any initial wavefunction $\Psi(t=0)$ in the Floquet states and immediately obtain the solution for arbitrary $t$. We can consider $\Omega = 2\pi/T$ to be the Floquet frequency. In the present work, $\Omega$ shall arise due to interaction of electrons with an electromagnetic wave.

An short proof is outlined here for completeness. ^{[20]}First, we attempt to decompose a general time evolution operator, $$U(t+nT,t_0)=U(t,t_0)[U(t_0+T,t_0)]^n,$$ the physical interpretation of which is quite clear from the proof, which simply follows: $$U(t+nT,t_0)=U(t+nT,t_0+nT)U(t_0+nT,t_0)=U(t,t_0)[U(t_0+T,t_0)]^n,$$ where in the last step, we've used periodicity, and also composed $n$ time translations by distances equal to the period, $T$. Define $$F_0\equiv U(t_0+T,t_0)\equiv e^{-iH_{F_0}t/\hbar}.$$ The eigenstates of this single-period operator are the Floquet modes evaluated initially, $|\Phi_\alpha(t_0)\rangle$, and the eigenvalues are $e^{-i\epsilon_\alpha T/\hbar}$. 

Furthermore, we can determine how these modes propagate in time. The derivation of this is slightly less obvious. Firstly, define $$P(t,t_0)\equiv U(t,t_0)e^{iH_{F_0} (t-t_0)}.$$ Notice $$P(t+T,t_0)=U(t+T,t_0)e^{iH_{F_0}(t+T-t_0)/\hbar}=U(t,t_0)U(t_0+T,t_0)e^{iH_{F_0}T}e^{iH_{F_0}(t-t_0)}=U(t,t_0)e^{iH_{F_0}(t-t_0)}=P(t,t_0),$$ establishing the periodicity of $P$. Finally, $$|\Phi_\alpha(t)\rangle=U(t,t_0)|\Phi_\alpha(t_0)\rangle=P(t,t_0)e^{-iH_{F_0}(t-t_0)}|\Phi_\alpha(t_0)\rangle.$$ Letting $H_{F_0}$ act on the Floquet modes to extract the quasienergies, the expression yields, $e^{-i\epsilon_\alpha(t-t_0)}P(t,t_0)|\Phi_\alpha(t_0)\rangle$. 

The Floquet modes thus propagate as $$|\Phi(t)_{\alpha}\rangle=P(t,t_0)|\Phi_\alpha(t_0)\rangle,$$ with the wavefunction evolving as $$|\Psi_\alpha(t)\rangle=e^{-i\epsilon_\alpha(t-t_0)}|\Phi_{\alpha}(t)\rangle.$$ Moreover, the Floquet modes inherit periodicity from $P(t,t_0)$.

####<center>Deriving the Floquet Hamiltonian

The Hamiltonian is periodic in time under the influence of monochromatic light, which is a reasonable approximation for a laser field. $^{[27]}$

Periodicity suggests a plausible simplification by disintegrating the Floquet modes into a Fourier basis, $$|\Phi_\alpha(t)\rangle=\sum_m e^{-im\Omega t} |\Phi_\alpha^m\rangle.$$ The sign of $m$, of course, is unimportant, as long as we're consistent.An inverse transform would help us go the other way.

As per the Floquet theorem, we take the ansatz $$|\Psi_{\alpha}(t)\rangle=\exp \left(-i \epsilon_{\alpha} t)\right |\Phi_{\alpha}t\rangle,$$ and substitute it into the time-dependent Schrodinger's equation. 

We define a `Floquet Hamiltonian operator'$^{[31]}$ as $$\mathcal{H}(t)=H_{0}(t)-i \frac{\partial}{\partial t}.$$ Note that $$i\frac{\partial |\Phi_\alpha(t)\rangle}{\partial t}=i\frac{\partial \exp(i\epsilon_\alpha t)|\Psi_\alpha(t)\rangle}{\partial t}=-\epsilon_\alpha\exp(i\epsilon_\alpha t)|\Psi_\alpha(t)\rangle+i\exp(i\epsilon_\alpha t)\frac{\partial|\Psi_\alpha(t)\rangle}{\partial t}
$$
$$=-\epsilon_\alpha\exp(i\epsilon_\alpha t)|\Psi_\alpha(t)\rangle+\exp(i\epsilon_\alpha t)H_0|\Phi_\alpha(t)\rangle=-\epsilon_\alpha|\Psi_\alpha(t)\rangle+H_0|\Phi_\alpha(t)\rangle$$. Therefore,
$$
\mathcal{H}(t)\left|\Phi_{ \alpha}(t)\right\rangle=\epsilon_{ \alpha}\left|\Phi_{ \alpha}(t)\right\rangle
$$
$$
\implies \sum_{n}(H_{0}(t) - n\Omega )e^{-in\Omega t}|\Phi_{\alpha}^{n}\rangle=\epsilon_{\alpha}\sum_{k}e^{-ik\Omega t}|\Phi_{\alpha}^{k}\rangle.
$$
Taking the Fourier transform with respect to time we get:
$$
\sum_{n}\frac{1}{T}(\int H_{0}(t) e^{i(m-n)\Omega t} dt - n\Omega \int e^{i(m-n)\Omega t} dt) |\Phi_{\alpha}^{n}\rangle = \epsilon_{\alpha} |\Phi_{\alpha}^{m}\rangle
$$
Making use of the fact that $H_{mn}\equiv\frac{1}{T}\int e^{i(m-n)\Omega t}dt H_0(t) $ and $\frac{1}{T}\int e^{i(m-n)\Omega t} dt=\delta_{mn}$, we can write this as:
$$\sum_{n}(H_{mn}-n\Omega\delta_{mn}) |\Phi_{\alpha}^{n}\rangle = \epsilon_{\alpha} |\Phi_{\alpha}^{m}\rangle$$
Therefore the Floquet Hamiltonian elements are given as:
$H^{F}_{mn} =H_{mn}-n\Omega\delta_{mn} $

To construct the $2\times2$ Floquet Hamiltonian in the paper, we used the Floquet indices $n=-1$ and $n=0$ corresponding to the valence and conduction bands respectively. The Hamiltonian is given as:
$$
H_{F}=\left(\begin{array}{cc}
\epsilon_{1}^{0}+\Omega & -i A^{*} v_{12}^{0} \\
i A v_{21}^{0} & \epsilon_{2}^{0}
\end{array}\right) \equiv \epsilon+\mathbf{d} \cdot \mathbf{\sigma}
$$

The derivation of this is not non-trivial.  Simply write $$H_0(t)=H_\text{original}(k+A(t))=H_\text{original}(k)+A(t)\partial_k H_\text{original}(k)+A(t)^2\partial_k^2  H_\text{original}(k)/2+\dots$$ for the Hamiltonian in presence of a field $$A(t)=i(Ae^{-i\Omega t}-A*e^{i\Omega t}).$$

We find the Floquet Hamiltonian $H_F$ by neglecting second order effects, and $H_F’$ by assuming that first order effects vanish, and neglecting third order effects.

For the first term in both cases,
$$H_{mn}^{(1)}=\int_0^Te^{i(m-n)\Omega t}H_\text{original}(k)dt=H_\text{original}(k)\delta_{mn},$$ where $\Omega T\equiv 1$.

Therefore, $$(H_F)_{mn}^{(1)}=H_{mn}^{(1)}-n\Omega\delta_{mn}=(H_\text{original}(k)-n\Omega)\delta_{mn}.$$

Note that $H_{mn}$ is a matrix. This makes $H_F$ a matrix of matrices. However, in the present scenario, we’ll primarily be interested in transitions between the valence and the conduction band, at a certain frequency. First order transitions occur at $\Omega$; second order transitions also require a $2\Omega$ terms. As a result, we’ll choose one Floquet band at $n=-1$ from the valence level $|u_1\rangle$, and another Floquet band at $n=0$ from the conduction level $|u_2\rangle$ for the former case. We thus have a “bijection” between Floquet and Bloch bands in each case. The Kronecker delta ensures that off-diagonal terms vanish. 

This gives for the first term of $H_F$, $$\begin{bmatrix}\epsilon_1^0+\Omega & 0\\ 0 & \epsilon_2^0\end{bmatrix},$$ where $\epsilon_1^0$ and $\epsilon_2^0$ were diagonal elements of our original Hamiltonian.

For the first term of $H_F’$, it means that we have $$\begin{bmatrix}\epsilon_1^0+2\Omega & 0\\ 0 & \epsilon_2^0\end{bmatrix}.$$ In this case, we picked $n=-2$ from the valence band and $n=0$ from the conduction band.

For the $A(t)$ order term in $H_F$, using the definition $\nu^0\equiv\partial_k H_\text{original}$, we have
$$H_F^{(2)}=\nu^0\int_0^T dt\,e^{i(m-n)\Omega t}i(Ae^{-i\Omega t}-A^{*}e^{i\Omega t}).$$
The integral integrates to a linear combination of $\delta(m-n+1)$ and $\delta(m-n-1)$, and thus is responsible for the off-diagonal terms of $H_F$. We must remember that we’re projecting onto a subspace with one Floquet band per Bloch band; as such, when $m=-1,n=0$, choose $\nu^0_{12}$. Similarly, choose $\nu^0_{21}$ when $m=0,n=-1$. This gives us, 
$$H_F^{(2)}=\begin{bmatrix}0 & -iA^{*}\nu^0_{12}\\ iA\nu_{21}^0 & 0\end{bmatrix}.$$

Similarly, we can write the $A(t)^2$ order term for $H_F’$ using $$H'^{(2)}_F=-i\partial_k\nu^0\int_0^T dt\,e^{i(m-n)\Omega t} (A^2e^{-2i\Omega t}+A^{*2}e^{2i\Omega t}+2|A|^2)/2.$$
This carries $\delta(m-n-2)$ and $\delta(n-m-2)$ terms. For the $m=-2,n=0$ matrix element, choose $(\partial_k\nu^0)_{12}$, and for $m=0,n=-2$, use $(\partial_k\nu^0)_{21}$ while calculating the metric elements. This yields,
$$H'^{(2)}_F=\begin{bmatrix}0 & -A^{*2}(\partial_k\nu^0)_{12}/2\\ -A^2(\partial_k\nu^0)_{21}/2 & 0\end{bmatrix}.$$

Finally, add up the computed components to conclude,

$$H_F=\begin{bmatrix} \epsilon_1^0+\Omega & -iA^*\nu^0_{12}\\ iA\nu_{21}^0 & \epsilon_2^0\end{bmatrix},$$

$$H_F’=\begin{bmatrix}\epsilon_1^0+2\Omega & -A^{*2}(\partial_k\nu^0)_{12}/2\\ -A^2(\partial_k\nu^0)_{21}/2 & \epsilon_2^0\end{bmatrix}.$$


###<center>1.3 Bloch's Theorem</center>
We quickly brush up this relatively straightforward topic from introductory solid state theory [24]. Given a particle in a spatially periodic potential, its wavefunction takes the form $$\psi(\mathbf{r})=e^{i\mathbf{k}\cdot\mathbf{r}}u(\mathbf{r}),$$
where $u(\mathbf{r})$ has the same periodicity in $\mathbf{r}$ as does the lattice itself. The periodicity results in the creation of "bands": distinct surfaces in energy vs $\mathbf{k}$-space, indexed by (without loss of generality, one) discrete index $n$. There is one band for each such $n$ indexing a bound state, over which the energies vary continuously as $E_n(\mathbf{k})$. The first Brillouin zone, a term frequently used, in this particular context will simply mostly be the period (or the unit cell) of the reciprocal lattice.
Also, a symmetry we shall frequently employ is that $E_n(\mathbf{k})=E_n(-\mathbf{k})$ [21,22]: Time reversal symmetry; flipping the sign of the "momentum" doesn't affect the energy.


###<center> 1.4 Berry Phases, Berry Connection and Berry Curvature

Topology is the formal study of shapes and spaces. The application of topology in Condensed Matter Physics is currently an exciting area of research and is leading to many new theoretical explanations of well-known phenomena such as the modern theory of polarisation as well as exciting developments on the frontiers of electronic devices and materials science. 

One of the major reasons that it is so interesting to study Condensed matter systems from a topological perspective is that these topological properties are not affected by small local changes, as they are related to the global structure of the system under study. Any impurities or defects in the material will not affect the result of a measurement of the topologically invariant property. 

To take a concrete example, the quantization of the Hall conductance has a topological origin, which makes the quantization of conductance very robust even at macro length scales and in the presence of impurities. This is because the topology of the system does not change continuously under deformations. $^{[35]}$

####<center>Berry Phase, Berry Connections and Berry Curvature

In classical and quantum mechanics, the geometric (Berry) phase is a phase difference acquired over the course of a cycle when a system is subjected to a cyclic adiabatic process. During this cyclic adiabatic process, the Hamiltonian is varied over some parameter space. The Berry phase is related to the geometry of this path in parameter space. Berry's original paper which also explained the Aharonov Bohm effect was a landmark paper, that in a way gave birth to the entire field of Topological Condensed Matter Physics.$^{[9]}$

Let us define some useful quantities, and then proceed to derive them using the adiabatic theorem.

The Berry connection in k-space is given as $^{[10],[11]}$
$${\mathcal{A}}_{n}=i\left\langle u_{n k}\left|\nabla_{\vec{k}}\right| u_{n k}\right\rangle$$.

This connection is a vector valued function which is analogous to a vector potential such as the magnetic vector potential in the Aharanov Bohm effect in Electromagnetism. However, we must be careful - this connection is defined over any parameter space, and in our case, it is the commonly used k-space over which our Hamiltonian varies adiabatically, rather than in real space.

The Berry connection when integrated over the closed path in parameter space gives us the Berry phase, which is experimentally observable. It plays a central role in describing topological phenomena, as we shall soon see.

The Berry curvature in k-space: $${\mathcal{F}}_{n}=\nabla_{\vec{k}} \times {\mathcal{A}}_{n}=i \epsilon_{ijk}\left\langle\partial_{k_{i}} u_{n k} \mid \partial_{k_{j}} u_{n k}\right\rangle$$
where $\epsilon_{ijk}$ is the Levi-Civita symbol.

It is simply the curl of the Berry-connection in k-space. It is analogous to a magnetic field, which is the curl of the magnetic vector potential. Like the magnetic field, the Berry connection is gauge invariant and is therefore physically relevant. More generally, the Berry connection is an anti-symmetric second-rank tensor derived from the Berry connection. In a three-dimensional parameter space the Berry curvature can be written in the pseudovector form given above.

####<center>Deriving the expression for the Berry Connection
We make use of the adiabatic theorem which states that under an adiabatic change of the Hamiltonian, a state starting in the nth eigenstate remains in the eigenstate. We consider the Hamiltonian cyclically varying over a contour in the parameter space $R(t)$.

We take the following ansatz: $^{[9],[14]}$

$\Psi_{n}(x, t)=\psi_{n}(x, t) e^{-\frac{i}{\hbar} \int_{0}^{t} E_{n}\left(t^{\prime}\right) d t^{\prime}} e^{i \gamma_{n}(t)}$.

The term $\theta_{n}(t) \equiv-\frac{i}{\hbar} \int_{0}^{t} E_{n}\left(t^{\prime}\right) d t^{\prime}$ is called the dynamical or trivial phase and does not depend on the geometry of the path in parameter space.

Plugging this into the time dependent Schrodinger equation and noting that $\dot{\theta_n} = E_n(t)$ and $H|\psi\rangle = E_n(t)$, we get:

$$
\frac{d \gamma_{n}}{d t}=i\left\langle\psi_{n} \mid \frac{\partial \psi_{n}}{\partial R}\right\rangle \frac{d R}{d t}
$$
$$
\gamma_{n}(t)=i \int_{0}^{t}\left\langle\psi_{n} \mid \frac{\partial \psi_{n}}{\partial R}\right\rangle \frac{d R}{d t^{\prime}} d t^{\prime}=i \int_{R_{i}}^{R_{f}}\left\langle\psi_{n} \mid \frac{\partial \psi_{n}}{\partial R}\right\rangle d R.
$$
If the Hamiltonian depends on more than one-parameter, we can generalise this to:
$\gamma_{n}(T)=i \oint\left\langle\psi_{n} \mid \nabla_{R} \psi_{n}\right\rangle \cdot d \mathbf{R}$/

This phase difference $\gamma$ is known as the Berry phase.

We notice that the integrand is simply what was defined earlier as the Berry connection.
$$
\gamma_{n}=\int_{\mathcal{C}} d \mathbf{R} \cdot \mathcal{A}_{n}(\mathbf{R}).
$$
Using Stoke's theorem, we can rewrite the Berry phase in terms of the Berry curvature as:
$\gamma_{n}=\int_{\mathcal{S}} d \mathbf{S} \cdot \mathcal{F}_{n}(\mathbf{R})$.

We have now established that there are quantities known as the Berry Phase, Berry connection and Berry curvature, and given them a geometrical meaning in parameter space. In order to now bring in the connection with topology, we state the Gauss-Bonnet theorem$^{[12,[16]]}$:

The Gauss–Bonnet theorem, or Gauss–Bonnet formula, is a relationship between surfaces in differential geometry. It connects the curvature of a surface (from geometry) to its Euler characteristic (from topology). The curvature is a local property which tells us how much a curve deviates from being a line or a surface deviates from being a plane. The Euler characteristic is topological and is related to the number of edges, vertices and faces a shape has. This will not change as the surface is slowly deformed though the curvature may change.

A common formulation is that for any compact, boundaryless two-dimensional Riemannian manifold($M$), the integral of the Gaussian curvature ($K$) over the entire manifold with respect to area is $2\pi$ times the Euler characteristic ($\chi_{M}$) of the manifold.
$$
\chi_{M}=\frac{1}{2 \pi} \oint_{M} K d S.
$$

This astonishing result implies that if we distort the surface and change the curvature at any location the same "total curvature" is maintained. Not only does the Euler characteristic not change under deformation - it is also an integer. 

To tie up the entire discussion so far in this section, we take the example of the Chern number, which is an example of a topologically invariant quantity.

The expression for the Chern number $C_n$ is
$$
C_{n}=\frac{1}{2 \pi} \oint_{\mathrm{BZ}} \mathcal{F}_{n}(\vec{k}) d \vec{k} \quad \quad \quad \quad C=\Sigma_{n} C_{n}
$$
where $n$ is the index of the filled bands.
The Chern number, a topologically invariant quantity, is always an integer as interpreted from the Gauss-Bonnet theorem$^{[13]}$ It is defined in terms of the Berry curvature. Therefore, a non-trivial Berry curvature will give rise to a non-zero Chern number, and this can be used to classify topological materials, or as we shall see, it can also be related to the quantized Hall conductance.


###<center> 1.5 The Quantum Hall Effect

Classically, the Hall effect is the production of a voltage difference across an conductor, transverse to an electric current in the conductor and to an applied magnetic field perpendicular to the current. The Hall conductance was found to be quantized. The quantization of the Hall conductance is of topological origin and is related to the Berry Connection and Curvature of the Bloch bands.$^{[18]}$ This explains the robust nature of the quantization of the transverse conductance. Here, we derive the Hall conductivity using the linear response of a quantum mechanical system. This is also known as the Kubo formula. $^{[15]}$
We consider the Hamiltonian of a periodic crystal which has translational symmetry. We consider a periodic potential (with primitive lattice vectors $a_{x} \hat{x}$ and $a_{y}\hat{y}$ potential (with primitive lattice vectors $a_{x} \hat{x}$ and $a_{y}\hat{y}$ ) with energy eigenstates of the Bloch form:
$$
\psi_{\mathrm{k}}(\mathrm{x})=u_{\mathrm{k}}(\mathrm{x}) e^{i \mathrm{k} \cdot \mathrm{x}} 
$$
$$
u_{\mathrm{k}}\left(\mathrm{x}+a_{x} \hat{\mathrm{x}}\right)=u_{\mathrm{k}}\left(\mathrm{x}+a_{y} \hat{\mathrm{y}}\right)=u_{\mathrm{k}}(\mathrm{x})
$$
We can now write the Hamiltonian, which depends on $k$, for the cell-periodic part of the Bloch function as $\operatorname{} H_{\mathrm{k}}=e^{-i \mathrm{k} \cdot \mathrm{x}} H e^{i \mathrm{kkk} \cdot \mathrm{x}}$ such that $H_{\mathrm{k}}\left|u_{\mathrm{k}}\right\rangle=E_{\mathrm{k}}\left|u_{\mathrm{k}}\right\rangle .$
The current operator is defined as the charge multiplied by the group velocity operator for a wavepacket of Bloch eigenstates.
$$
\mathbf{J}=\frac{e}{\hbar} \frac{\partial H_{\mathbf{k}}}{\partial \mathbf{k}}
$$
Before we proceed, we assume that the Hamiltonian for the system neglects electron-electron interactions, and that the Fermi energy $E_{F}$ lies in a gap between bands.  

We now use the Kubo forumla (linear response theory)$^{[17]}$for conductivity in the energy eigenstate $|m\rangle$
(where $|n\rangle$ are eigenstates of the Hamiltonian with energy eigenvalues $E_{n}$).
$$
\sigma_{x y}=i \hbar \sum_{n \neq m} \frac{\left\langle m\left|J_{y}\right| n\right\rangle\left\langle n\left|J_{x}\right| m\right\rangle-\left\langle m\left|J_{x}\right| n\right\rangle\left\langle n\left|J_{y}\right| m\right\rangle}{\left(E_{n}-E_{m}\right)^{2}}
$$
In order to describe conductivity in the many-body ground state, we must modify it and take the continuum limit$^{[19]}$
$$
\sigma_{x y}=\frac{i e^{2}}{\hbar} \sum_{E_{\alpha}<E_{F}<E_{\beta}} \int \frac{d^{2} \mathbf{k}}{(2 \pi)^{2}} \frac{d^{2} \mathbf{k}^{\prime}}{(2 \pi)^{2}} \mathcal{S}\left(u_{\mathbf{k}}^{\alpha}, u_{\mathbf{k}^{\prime}}^{\beta}\right)
$$
where
$$
\begin{array}{r}
\mathcal{S}\left(u_{\mathbf{k}}^{\alpha}, u_{\mathbf{k}^{\prime}}^{\beta}\right)=\frac{\left\langle u_{\mathbf{k}}^{\alpha}\left|\frac{\partial H_{\mathbf{k}}}{\partial k^{y}}\right| u_{\mathbf{k}^{\prime}}^{\beta}\right\rangle\left\langle u_{\mathbf{k}^{\prime}}^{\beta}\left|\frac{\partial H_{\mathbf{k}}}{\partial k^{x}}\right| u_{\mathbf{k}}^{\alpha}\right\rangle}{\left(E_{\beta}\left(\mathbf{k}^{\prime}\right)-E_{\alpha}(\mathbf{k})\right)^{2}} \\
-\frac{\left\langle u_{\mathbf{k}}^{\alpha}\left|\frac{\partial H_{\mathbf{k}}}{\partial k^{x}}\right| u_{\mathbf{k}^{\prime}}^{\beta}\right\rangle\left\langle u_{\mathbf{k}^{\prime}}^{\beta}\left|\frac{\partial H_{\mathbf{k}}}{\partial k^{y}}\right| u_{\mathbf{k}}^{\alpha}\right\rangle}{\left(E_{\beta}\left(\mathbf{k}^{\prime}\right)-E_{\alpha}(\mathbf{k})\right)^{2}}
\end{array}
$$
Here, the integrals are over the Brillouin zone, while the sum is over pairs of states $u_{\mathrm{k}^{\prime}}^{\beta}, u_{\mathrm{k}}^{\alpha}$ above and below the Fermi energy respectively.We use the completeness relation and the fact that $$\langle u_{\mathbf{k}}^{\alpha}|\frac{\partial H_{\mathbf{k}}}{\partial k^{y}}|u_{\mathbf{k^{\prime}}}^{\beta}\rangle = -(E_{\beta}(k^{\prime})-E_{\alpha}(k))\langle \frac{\partial u_{\mathbf{k}}^{\alpha}}{\partial k^{y}} | u_{\mathbf{k^{\prime}}}^{\beta}\rangle $$
 
to get:
$$
\sigma_{x y}=\frac{i e^{2}}{\hbar} \sum_{E_{\alpha}<E_{F}} \int \frac{d^{2} \mathbf{k}}{(2 \pi)^{2}}\left[\left\langle\frac{\partial u_{\mathbf{k}}^{\alpha}}{\partial k^{y}} \mid \frac{\partial u_{\mathbf{k}}^{\alpha}}{\partial k^{x}}\right\rangle-\left\langle\frac{\partial u_{\mathbf{k}}^{\alpha}}{\partial k^{x}} \mid \frac{\partial u_{\mathbf{k}}^{\alpha}}{\partial k^{y}}\right\rangle\right]
$$
We now observe that the expression for the Berry curvature has popped up in the modified Kubo formula. Thus, we have related the Hall conductivity to a fundamental topological invariant of the system. 
The Berry connection over the Brillouin zone is:
$$
\mathcal{A}_{j}(\alpha, \mathbf{k})=-i\left\langle u_{\mathbf{k}}^{\alpha}\left|\frac{\partial}{\partial k^{j}}\right| u_{\mathbf{k}}^{\alpha}\right\rangle
$$
The Berry curvature corresponding to this connection is
$$
\begin{aligned}
\mathcal{F}_{i j}(\alpha) &=\frac{\partial \mathcal{A}_{j}(\alpha, \mathbf{k})}{\partial k^{i}}-\frac{\partial \mathcal{A}_{i}(\alpha, \mathbf{k})}{\partial k^{j}} \\
&=-i\left\langle\frac{\partial u_{\mathbf{k}}^{\alpha}}{\partial k^{i}} \mid \frac{\partial u_{\mathbf{k}}^{\alpha}}{\partial k^{j}}\right\rangle+i\left\langle\frac{\partial u_{\mathbf{k}}^{\alpha}}{\partial k^{j}} \mid \frac{\partial u_{\mathbf{k}}^{\alpha}}{\partial k^{i}}\right\rangle
\end{aligned}
$$
Using this, we can write the expression for Hall conductivity in terms of the Berry curvature:
$$\sigma_{x y}=\frac{e^{2}}{\hbar} \sum_{\alpha} \int \frac{d k^{x} d k^{y}}{(2 \pi)^{2}} \mathcal{F}_{x y}(\alpha)$$
\newline
The integral of the Berry curvature $\mathcal{F}_{x y}(\alpha)$ over the Brillouin zone is $2 \pi C_{\alpha},$ with $C_{\alpha}$ an exact integer, referred to as the Chern number,as discussed in the previous section. Therefore, we have proved that the quantization of conductance in units of $\frac{e^2}{h}$ is related to the Berry curvature over the Brillouin zone.$^{[19]}$
$$
\sigma_{x y}=\frac{e^{2}}{2 \pi \hbar} \sum_{\alpha} C_{\alpha} \in \frac{e^{2}}{2 \pi \hbar} \mathbb{Z}
$$

In the context of the nonlinear Quantum Hall effect in 2D materials, we find the nonlinear Hall response by applying the linear response theory to the non-equilibrium steady states as:
$$\sigma_{x y}=\int_{\mathrm{BZ}} d^{2} k \sum_{i} f_{i}\left(\nabla \times \tilde{a}_{i}\right)_{z}$$


###<center>1.5 The 1D chain; Tight binding model</center>


First note that the operator $\hat{N} = c^\dagger_i c_i$ acting on a state with the orbital on atom $i$ occupied (or unoccupied) gives us the number of electrons that occupy that orbital as its eigenvalue (for fermions this value may only be 0 or 1). If the state has orbital $i$ occupied, the operator $\hat{N}$ does nothing but annihilate the electron and and subsequently create it, giving us an identical state. This corresponds to  an eigenvalue of 1. If the state has orbital $i$ unoccupied, the action of $c_i$ is to simply make the state vanish, corresponding to an eigenvalue of 0. 

Thus a Hamiltonian ascribing an energy $V_i$ to every occupied orbital $i$ may be written as follows:

\begin{equation}
    H_0 = \sum_{i} V_i  c^\dagger_i c_i
\end{equation}

We can also add interaction terms of the form $c^\dagger_i  c_j$ to account for the interaction energy between sites $i$ and $j$. Terms of this form annihilate an electron at one site and create it at another, and are hence often called <i>hopping</i> terms. Assigning an interaction energy $T_{ij}$ for every pair of sites in the system gives us $^{[38]}$:

\begin{equation}
    H_0 = \sum_{i} V_i  c^\dagger_i c_i + \sum_{ij} T_{ij} \left( c^\dagger_i c_j + c^\dagger_j c_i \right)
\end{equation}

The interaction term has been written in the form $c^\dagger_i c_j + c^\dagger_j c_i$ to ensure hermiticity of the Hamiltonian. 

Now consider the 1D chain model. It is a model described by a set of atoms like the ones below $^{[36]}$:




oneDchain_color.png

The system consists of two Bravais sublattices A and B, and the two different kinds of sites may have different occupation energies. The odd sites have a lesser occupation energy than the even (say). This gives us the following non interacting Hamiltonian: (the energies are chosen to oscillate around 1 in arbitrary units)

\begin{equation}
    H_0 = \sum_i (1 + Q_{1}(-1)^{i}) c_{i}^{\dagger} c_{i} = N + \sum_i Q_{1}(-1)^{i} c_{i}^{\dagger} c_{i}
\end{equation}

Since the number of electrons ($N$) in the 1-D chain remains constant, we may ignore this term and write a cleaner Hamiltonian as below:

\begin{equation}
    H_0 = \sum_i Q_{1}(-1)^{i} c_{i}^{\dagger} c_{i}
\end{equation}

Now let us add interaction energies. For nearest neighbor interactions, we expect an A-B interaction (single bond) to be weaker than a B-A interaction (double bond). This gives us the following Hamiltonian:

\begin{equation}
    H_0 =\sum_{i} \frac{1+Q_{2}(-1)^{i}}{2}\left( c_{i}^{\dagger} c_{i+1}+\text {h.c.}\right) + \sum_i Q_{1}(-1)^{i} c_{i}^{\dagger} c_{i}
\end{equation}

We know introduce a term to break $T$ symmetry. Let the magnitude of interaction between next-nearest neighbors (that is, A-A and B-B) be $Q'_1$, having opposite signs for different sublattices. This term introduces a factor of $\sin k$ in the fourier space Hamiltonian, which is why it breaks time-reversal symmetry. This gives us, finally:

\begin{equation}
\begin{aligned}
H_{0}=& \sum_{i} \frac{1+Q_{2}(-1)^{i}}{2}\left( c_{i}^{\dagger} c_{i+1}+h . c .\right) \\
&+Q_{1}(-1)^{i} c_{i}^{\dagger} c_{i}+Q_{1}^{\prime}(-1)^{i}\left( c_{i}^{\dagger} c_{i+2}+\text {h.c.}\right)
\end{aligned}
\end{equation}

###<center>2.1 The Shift Current</center>


We list relations from previous sections which will find relevance here. Under an electric field $Ee^{i\Omega t}+E^*e^{-i\Omega t}$, Firstly, the Floquet Hamiltonian, written in terms of the corresponding vector potential $A$ and its complex conjugate $A^*$, as
$$H_F=\begin{bmatrix}\epsilon_1^0+\Omega && -iA^*\nu_{12}^0\\ iA\nu_{21}^0 &&\epsilon_2^0\end{bmatrix}=\epsilon+\mathbf{d}\cdot\mathbf{\sigma}.$$
It is fit to talk about the physical interpretation of this Hamiltonian here. We have the unperturbed energy eigenvalues along the diagonal. The "ground state" (valence band) is dressed with a photon of energy $\Omega$. Interaction with the electromagnetic field further creates the off-diagonal terms. 
The DC current operator is given by the derivative of the energy operator (Hamiltonian) with respect to the momentum ($\partial E/\partial p=v_G$, the group velocity, which is in turn proportional to the current):
$$\tilde \nu=\frac{\partial H_F}{\partial k}=\begin{bmatrix}\frac{\partial \epsilon_1^0}{\partial k} && -iA^*\left(\frac{\partial\nu^0}{\partial k}\right)_{12}\\ iA\left(\frac{\partial\nu^0}{\partial k}\right)_{21} &&\frac{\partial \epsilon_2^0}{\partial k}\end{bmatrix}=\begin{bmatrix}\nu_{11}^0 && -iA^*\left(\frac{\partial\nu^0}{\partial k}\right)_{12}\\ iA\left(\frac{\partial\nu^0}{\partial k}\right)_{21} &&\nu_{22}^0\end{bmatrix}=b_0+b\cdot\sigma.$$
Lastly, the lesser Green's function for the current situation, as derived in the previous section in terms of the self energy, which simplifies to a projector, $$\Sigma^<=i\Gamma(1+\sigma_z)/2=i\Gamma\begin{bmatrix}1&&0\\0&&0\end{bmatrix}=i\Gamma P_{11},$$ $$G^<=\frac{(\omega-\epsilon+i\Gamma/2+d\cdot\sigma)\Sigma^<(\omega-\epsilon-i\Gamma/2+d\cdot\sigma)}{[(\omega-\epsilon+i\Gamma/2)^2-d^2][(\omega-\epsilon-i\Gamma/2)^2-d^2]}=i\Gamma\frac{(\alpha+d\cdot\sigma)P_{11}(\alpha^*+d\cdot\sigma)}{[\alpha^2-d^2][\alpha^{*2}-d^2]},$$ where we have identified $\alpha=\omega-\epsilon+i\Gamma/2$.
With the lesser Green's function being proportional to a occupation density matrix and all, we have $$J=-i\text{Tr}(\tilde\nu G^<)=-i\text{Tr}[(b_0+b\cdot\sigma)G^<].$$

At this point, it's useful to note that $P_{11}\sigma_x=P_{12},P_{11}\sigma_y=-iP_{12},P_{11}\sigma_z=P_{11},P_{11}\mathbb{I}=P_{11}$, which implies that $$(\alpha+d\cdot\sigma)P_{11}(\alpha^*+d\cdot\sigma)=\begin{bmatrix}\alpha+d_z && d_x-id_y\\ d_x+i d_y &&\alpha-d_z\end{bmatrix}\begin{bmatrix}\alpha^*+d_z && d_x-id_y\\ 0&&0\end{bmatrix}=\begin{bmatrix}|\alpha+d_z|^2 && (\alpha+d_z)(d_x-id_y)\\ (\alpha^*+d_z)(d_x+id_y)&&d_x^2+d_y^2\end{bmatrix}.$$
Note that $P_{ij}\equiv |i\rangle\langle j|$.

Now, we must evaluate $$J=\text{Tr}_{\omega,k}\{\frac{\Gamma}{(\alpha^2-d^2)(\alpha^{*2}-d^2)}\text{Tr}_2\left(\begin{bmatrix}b_0+b_z&&b_x-ib_y\\b_x+ib_y&&b_0-b_z\end{bmatrix}\begin{bmatrix}|\alpha+d_z|^2 && (\alpha+d_z)(d_x-id_y)\\ (\alpha^*+d_z)(d_x+id_y)&&d_x^2+d_y^2\end{bmatrix}\right)\}.$$ A reminder: Remember that the trace in the original definition is not only the trace of the 2 by 2 matrix; we also need to trace over the $\omega$ and $k$ spaces, through a straightforward integration; thus the subscripts.

The $2\times 2$ trace evaluates to $$(b_0+b_z)|\alpha+d_z|^2+(b_x-ib_y)(d_x+id_y)(\alpha^*+d_z)+(b_x+ib_y)(d_x-id_y)(\alpha+d_z)+(b_0-b_z)(d_x^2+d_y^2)=(b_0+b_z)[(\omega-\epsilon+d_z)^2+\Gamma^2/4]+(b_0-b_z)(d_x^2+d_y^2)+2(b_xd_x+b_yd_y)(\omega-\epsilon+d_z)+\Gamma (b_x d_y-b_y d_x)=(b_0+b_z)[(\omega-\epsilon)^2+2(\omega-\epsilon)d_z+d_z^2+\Gamma^2/4]+(b_0-b_z)(d^2-d_z^2)+\Gamma (b_x d_y-b_y d_x)+2(b_xd_x+b_yd_y)(\omega-\epsilon+d_z).$$

Observe the denominator of the prefactor, $$(\alpha^2-d^2)(\alpha^{*2}-d^2)=(\alpha+d)(\alpha^*+d)(\alpha-d)(\alpha^*-d)=[(\omega-\epsilon)^2+d^2+\Gamma^2/4-2(\omega-\epsilon)d][(\omega-\epsilon)^2+d^2+\Gamma^2/4+2(\omega-\epsilon)d]=[(\omega-\epsilon)^2+d^2+\Gamma^2/4]^2-4(\omega-\epsilon)^2 d^2,$$ an even function of $(\omega-\epsilon)$. Since we shall be tracing over $\omega$, the integral of functions odd in $(\omega-\epsilon)$ will vanish; appropriately remove the linear terms from the numerator to obtain, $$(b_0+b_z)[(\omega-\epsilon)^2+d_z^2+\Gamma^2/4]+(b_0-b_z)(d^2-d_z^2)+\Gamma (b_x d_y-b_y d_x)+2(b_xd_x+b_yd_y)d_z=(b_0+b_z)[(\omega-\epsilon)^2+d^2+\Gamma^2/4]-2b_z(d^2-d_z^2)+\Gamma (b_x d_y-b_y d_x)+2(b_xd_x+b_yd_y)d_z.$$

Before we forget its existence, let's put the $\Gamma$ prefactor back, to get a complete expression for the non-vanishing (under trace) part of the numerator, as determined up till now: $$\Gamma\{(b_0+b_z)[(\omega-\epsilon)^2+d^2+\Gamma^2/4]-2b_z(d^2-d_z^2)+\Gamma (b_x d_y-b_y d_x)+2(b_xd_x+b_yd_y)d_z\}.$$

We integrate term by term. Single out the relevant portion of the first term, $$\int_{-\infty}^\infty \frac{d\omega}{2\pi}\frac{(\omega-\epsilon)^2+d^2+\frac{\Gamma^2}{4}}{[(\omega-\epsilon)^2+d^2+\Gamma^2/4-2(\omega-\epsilon)d][(\omega-\epsilon)^2+d^2+\Gamma^2/4+2(\omega-\epsilon)d]}=\int_{-\infty}^\infty \frac{d\omega}{2\pi}\frac{1}{2}\left(\frac{1}{(\omega-\epsilon)^2+d^2+\Gamma^2/4-2(\omega-\epsilon)d}+\frac{1}{(\omega-\epsilon)^2+d^2+\Gamma^2/4+2(\omega-\epsilon)d}\right)=\int_{-\infty}^\infty \frac{d\omega}{2\pi}\frac{1}{2}\left(\frac{1}{(\omega-\epsilon-d)^2+\Gamma^2/4}+\frac{1}{(\omega-\epsilon+d)^2+\Gamma^2/4}\right)=\frac{1}{2\pi}\frac{1}{2}\left(\frac{2\pi}{\Gamma}+\frac{2\pi}{\Gamma}\right)=\frac{1}{\Gamma}.$$

The rest of the numerator is completely independent of $\omega$. We thus need the value of the following terrible integral (with $(\omega-\epsilon)\to x$): $$\int_{-\infty}^\infty dx \frac{1}{2\pi [(x-d)^2+\Gamma^2/4][(x+d)^2+\Gamma^2/4]}=\frac{1}{2(d^2+\Gamma^2/4)\Gamma}.$$ We delegated the honour of doing this to Mathematica, assuming all physical variables to be real (in the relevant code). This may even be derived through a simple application of partial fractions. 

Put together the terms with the values of the integrals: $$J=\text{Tr}_k\left((b_0+b_z)-\frac{b_z(d^2-d_z^2)}{d^2+\Gamma^2/4}+\frac{\Gamma}{2} \frac{(b_x d_y-b_y d_x)}{(d^2+\Gamma^2/4)}+\frac{(b_xd_x+b_yd_y)d_z}{d^2+\Gamma^2/4}\right)=\text{Tr}_k\left(b_0+\frac{b_z(\Gamma^2/4+d_z^2)}{d^2+\Gamma^2/4}+\frac{\Gamma}{2} \frac{(b_x d_y-b_y d_x)}{(d^2+\Gamma^2/4)}+\frac{(b_xd_x+b_yd_y)d_z}{d^2+\Gamma^2/4}\right).$$

Going as per the convention used in the paper, label $$j_3\equiv b_0+\frac{b_z(\Gamma^2/4+d_z^2)}{d^2+\Gamma^2/4},$$$$j_1\equiv \Gamma(b_xd_y-b_yd_x)/[2(d^2+\Gamma^2/4)],$$$$ j_2\equiv (b_xd_x+b_yd_y)d_z/(d^2+\Gamma^2/4).$$

Write $$b_xd_y-d_xb_y=\mathbf{Im}[(d_x+id_y)(b_x-ib_y)]=|A|^2\mathbf{Im}[\nu_{21}^0\left(\frac{\partial \nu^0}{\partial k}\right)_{12}].$$ Now, $$\left(\frac{\partial \nu^0}{\partial k}\right)_{12}=\langle u_1|\frac{\partial \nu^0}{\partial k}|u_2\rangle=\partial_k\langle u_1|\nu^0|u_2\rangle-\langle\partial_k u_1|\nu^0| u_2 \rangle-\langle u_1|\nu^0| \partial_k u_2 \rangle=\partial_k\nu_{12}^0-\langle\partial_k u_1|u_1\rangle\langle u_1|\nu^0| u_2 \rangle-\langle\partial_k u_1|u_2\rangle\langle u_2|\nu^0| u_2 \rangle-\langle u_1|\nu^0| u_1\rangle\langle u_1|\partial_k u_2 \rangle-\langle u_1|\nu^0| u_2\rangle\langle u_2|\partial_k u_2 \rangle=\partial_k\nu_{12}^0-\langle\partial_k u_1|u_1\rangle \nu^0_{12}-\langle\partial_k u_1|u_2\rangle\nu_{22}^0-\nu_{11}^0\langle u_1|\partial_k u_2 \rangle-\nu_{12}^0\langle u_2|\partial_k u_2 \rangle.$$

Use $$a_i=-i\langle u_i|\partial_k u_i\rangle=i \langle\partial_k u_i|u_i\rangle$$ (since differentiation with respect to momentum (position) flips sign under "parity/reversal"). From properties of the two-band model, we also know that $$\langle u_1|\partial_k u_2\rangle=-\frac{\nu_{12}^0}{\epsilon_1^0-\epsilon_2^0}, \langle \partial_k u_1| u_2\rangle=-\frac{\nu_{12}^0}{\epsilon_2^0-\epsilon_1^0}$$ 
The second can be obtained from the first by switching indices and taking the hermitian conjugate of both sides, and vice versa. To obtain one of them, note that $$H_0=\epsilon_1^0|u_1\rangle\langle u_1|+\epsilon_2^0|u_2\rangle\langle u_2|\implies \partial_k H_0=(\partial_k\epsilon_1^0)|u_1\rangle\langle u_1|+\epsilon_1^0 (\partial_k |u_1\rangle)\langle u_1| +\epsilon_1^0 |u_1\rangle\partial_k(\langle u_1|)+(\partial_k\epsilon_2^0)|u_2\rangle\langle u_2|+\epsilon_2^0 (\partial_k |u_2\rangle)\langle u_2| +\epsilon_2^0 |u_2\rangle\partial_k(\langle u_2|)\implies \langle u_1|\partial_k H_0|u_2\rangle=\epsilon_1^0\partial_k(\langle u_1|)|u_2\rangle+\epsilon_2^0\langle u_1|\partial_k|u_2\rangle\implies \nu^0_{12}=\epsilon_1^0\langle u_1|\partial_k^\dagger|u_2\rangle+\epsilon_2^0\langle u_1|\partial_k|u_2\rangle=\langle u_1|\partial_k|u_2\rangle (\epsilon_2^0-\epsilon_1^0)\implies \langle u_1|\partial_k u_2\rangle=-\frac{\nu_{12}^0}{\epsilon_1^0-\epsilon_2^0}.$$

With all of these substitutions, and factoring out $\nu_{12}^0$, we get $$\left(\frac{\partial \nu^0}{\partial k}\right)_{12}=\nu_{12}^0[\partial_k\ln\nu_{12}^0+ia_1-ia_2+\frac{\nu_{11}^0-\nu_{22}^0}{\epsilon_1-\epsilon_2}].$$ Thus define $$\phi_{12}\equiv\mathbf{Im}(\ln\nu_{12}^0), \, R_k=\partial_k \phi_{12}+a_1-a_2.$$ The term $R_k$ is called the shift vector, and is gauge-invariant. Moreover, $\phi_{12}$, being the phase of $\nu_{12}^0$, equals $-\phi_{21}$. This definition helps us nicely extract the useful chunk of imaginary terms in the relevant matrix element for the derivative. Therefore, $$d_xb_y-d_yb_x=|A\nu_{12}^0|^2R_k.$$

Quickly note that $$d^2=d_z^2+(d_x+id_y)(d_x-id_y)=d_z^2+(-iA^*\nu_{12}^0)(iA\nu_{12}^0)=d_z^2+|A\nu_{12}^0|^2.$$ Now we finally have the means to claim $$j_1=|A\nu^0_{12}|^2R_k\frac{\Gamma/2}{d_z^2+|A\nu^0_{12}|^2+\Gamma^2/4}.$$ Furthermore, if we assume that $$d_z\gg |A\nu_{12}^0|,\Gamma,$$ we get a sharply peaked curve, which may be approximated by a constant multiple of $\delta(d_z)$. To find the constant factor, one must integrate the former expression over all $d_z$; this is straightforward to do analytically, giving arctangents to put limits into. Finally, we obtain $$j_1\approx\frac{\pi|E|^2}{\Omega^2}\frac{\Gamma/2}{\sqrt{\Gamma^2/4+|E\nu^0_{12}|^2/\Omega^2}}\delta(d_z)|\nu_{12}^0|^2R_k.$$ Assuming small electric fields, $|A\nu_{12}^0|\ll \Gamma$, this simplifies further still, to $$j_1\approx\pi |E\nu_{12}^0/\Omega|^2R_k\delta(d_z).$$

Now for $j_2$: $$d_xb_x+d_yb_y=\Re[(d_x+id_y)(b_x-ib_y)]=|A|^2\Re[\nu_{21}^0\left(\frac{\partial \nu^0}{\partial k}\right)_{12}].$$ Hence, $$j_2=|A|^2\Re[\nu_{21}^0\left(\frac{\partial \nu_0}{\partial k}\right)_{12}]d_z/(d^2+\Gamma^2/4).$$ This will be proportional to the term $$\frac{\nu_{11}^0-\nu_{22}^0}{\epsilon_1-\epsilon_2}$$ from the olden expression. Now, eigen-energies are even in $k$, so their derivatives, $\nu_{ii}$ are odd, and the integral over all $k$ of these quantities vanishes. This forces $j_2=0$. Also, $$j_3=\text{Coefficient}\times b_z+b_0; b_z\propto (\nu_{11}-\nu_{22}); b_0\propto (\nu_{11}+\nu_{22})$$ again, time reversal symmetry ($k$-parity) forces $j_3$ to vanish. With all our approximations, we get, by integrating $j_1$ over all $k$, a current proportional to the square of field,
$$J=\frac{\pi|E|^2}{\Omega^2}\int dk \delta(\epsilon_1^0-\epsilon_2^0+\hbar\omega)|\nu_{12}^0|^2[\partial_k\phi_{12}+a_1-a_2].$$

Higher order corrections are captured by $\Gamma/\sqrt{4|E\nu_{12}^0|^2/\Omega^2+\Gamma^2}$, as visible from before we brought in the Dirac delta approximation. This leads, for large intensities and thus large $|E|$, to currents proportional to $|\Gamma E|$, instead of $|E|^2$, as is the case for small intensities.

From the Floquet formalism, as discussed previously, we want to find the current $J(2\Omega)\propto e^{-2i\Omega t}$; thus write $J(2\Omega)=-i\nu'_{12}(G^<)_{21}$, and choose $\nu'_{12}$ so as to enforce an $e^{-2i\Omega t}$ proportionality. When we choose bands labelled $n=-1$ and $n=0$ (i.e., Hamiltonian $H_F$, we have an exponential $e^{-i\Omega t}$ outside. On the other hand, when we use bands labelled $n=-2$ and $n=0$, i.e., Hamiltonian $H_F'$, we want a non-oscillating $\partial H_0/\partial k$ term accompanying it, since $J_{mn}$ already carries an $e^{-2i\Omega t}$ dependence. for the latter. We have assumed that $|A\nu_{12}^0|\ll d^2+\Gamma^2/4$, and a similar relation for the second case, where $d_x^2+d_y^2=\frac{|A|^4\|\partial_k\nu_{12}^0|^2}{4}$: In essence, as in the previous section, we're working under the assumption that the field is small.

Put $E=A/\Omega$ to finally obtain

Hence, the terms within the brackets simplify as $$(\frac{1}{2(d_z'+i\Gamma/2)}-\frac{1}{d_z+i\Gamma/2})=(\frac{1}{\epsilon_1^0-\epsilon_2^0+2\Omega+i\Gamma}-\frac{2}{\epsilon_1^0-\epsilon_2^0+\Omega+i\Gamma}).$$ Focus on the contribution close to resonance. Again, as in the previous section, we get constants times Dirac's delta; assuming small $\Gamma$, the terms in the bracket, are, up to a multiplicative constant, $[\delta(\epsilon_1^0-\epsilon_2^0+2\Omega)-2\delta(\epsilon_1^0-\epsilon_2^0+\Omega)]$. To find the constant, equate the integrals over all $\Omega$ for both sides. To do the complex integral, use the Cauchy's integral formula on a contour going from $-\infty\to\infty\to\infty+2i\Gamma\to-\infty+2i\Gamma\to-\infty$, along with symmetry, to get $-i\pi$ for the constant of proportionality. Hence, finally,

$$J(2\Omega)\approx \frac{\pi E^2}{2\Omega^2}\int dk |\nu_{12}^0|^2R_k[\delta(\epsilon_1^0-\epsilon_2^0+2\Omega)/2-\delta(\epsilon_1^0-\epsilon_2^0+\Omega)].$$

###<center>2.3 Third Order Response

The third order response is present in all materials but is generally not significant compared to first order and second order responses as it is several orders of magnitude smaller. In crystals where there is inversion symmetry, $\chi^{(2)}$ vanishes, and third order nonlinear responses become significant.

In our paper, we study the third order nonlinear responses described by:

$$ J_i = \chi_{ij}^{(3)}E(\omega)E(-\omega)E_j.$$ 

where the current $J_i$ is induced by the static electric field $E_j$ in the presence of the pump laser light of frequency $\omega$.

In general, the expression for dc current in the presence of a monochromatic laser with optical field E and a static electric field $E_0$ is given as:$^{[26]}$

$$
\mathbf{J}_{\mathrm{dc}}=\sigma_{\mathrm{dark}}^{(1)} \mathbf{E}_{0}+\sigma_{\mathrm{dark}}^{(2)} \mathbf{E}_{0}+\sigma_{b p v e}^{(2)} \mathbf{E}^{2}+\sigma_{p h}^{(3)} \mathbf{E}^{2} \mathbf{E}_{0}+\cdots.
$$
The first and second terms are the linear and quadratic dc conductivity in the absence of illumination. The third is the Bulk Photovoltaic Effect and the fourth is the photoconductivity, which is the the intensity dependent dc conductivity. The photoconductivty is a fourth rank tensor, which is related to the third order nonlinear susceptibility $\chi^{(3)}$ and its divergences. ^{[28]}$

The paper focuses on the nonlinear Quantum Hall effect in 2D materials. To find the nonlinear Hall response we apply the linear response theory (discussed in a previous section) to the non-equilibrium steady states and expanding it to the second power of $|E(\omega)|.$

We have already established that the expression for the Hall conductance is given by:
$$\sigma_{x y}=\int_{\mathrm{BZ}} d^{2} k \sum_{i} f_{i}\left(\nabla \times \tilde{a}_{i}\right)_{z}.$$

Here the Berry connection $\tilde{a_i}$ is defined for the Floquet states that describe the nonequilibrium steady states and $f_i$ is the occupation of the $i$th Floquet state.

Using our previous model for the two-band Floquet Hamiltonian discussed before, we use the standard spinor representation to get the eigenstates as:
$$
u_{1}=\left(\begin{array}{c}
\cos \frac{\theta}{2} \\
\sin \frac{\theta}{2} e^{i \phi}
\end{array}\right), \quad u_{2}=\left(\begin{array}{c}
-\sin \frac{\theta}{2} \\
\cos \frac{\theta}{2} e^{i \phi}
\end{array}\right)
$$

where cos$(\theta)$ = $\frac{d_z}{d}$  and $\phi=\frac{d_x}{d_y}$.

Now, we express the Floquet states in terms of the equilibirum eigenstates as:
$$|u_1\rangle =cos(\theta/2)|u_{1eq}\rangle + e^{i\phi}sin(\theta/2)|u_{2eq}\rangle $$
$$|u_2\rangle =-sin(\theta/2)|u_{1eq}\rangle + e^{i\phi}cos(\theta/2)|u_{2eq}\rangle $$.

To find the Berry connection  ($i\langle u_i|\nabla_k|u_i\rangle$) we first find the derivative of the states with respect to k as:

$$ \nabla_k |u_1\rangle = \frac{\partial |u_1\rangle}{\partial \theta}\nabla_k\theta + \frac{\partial |u_1\rangle}{\partial \phi}\nabla_k\phi  + cos(\theta/2)|\nabla u_{1eq}\rangle + e^{i\phi}sin(\theta/2)|\nabla_ku_{2eq}\rangle$$ 

$$ \nabla_k |u_2\rangle = \frac{\partial |u_2\rangle}{\partial \theta}\nabla_k\theta + \frac{\partial |u_2\rangle}{\partial \phi}\nabla_k\phi  + e^{i\phi}cos(\theta/2)|\nabla u_{2eq}\rangle -sin(\theta/2)|\nabla_ku_{1eq}\rangle$$.

We then take their inner-products with $\langle u_1|$ and $\langle u_2|$ to get the Berry connections. We note that the term with the derivative with respect to $\theta$ drops out because $\frac{\partial|u_1\rangle}{\partial \theta}= |u_2\rangle$ which is orthogonal to $|u_1\rangle.$ and $\frac{\partial|u_2\rangle}{\partial \theta}= |u_1\rangle$ which is orthogonal to $|u_2\rangle.$

Therefore our final expression for the Berry connection is:
$$
\begin{array}{l}
\tilde{\mathbf{a}}_{1}=\frac{1}{2}(1-\cos \theta) \nabla \phi+\cos ^{2} \frac{\theta}{2} \mathbf{a}_{1}+\sin ^{2} \frac{\theta}{2} \mathbf{a}_{2} \\
\tilde{\mathbf{a}}_{2}=\frac{1}{2}(1+\cos \theta) \nabla \phi+\sin ^{2} \frac{\theta}{2} \mathbf{a}_{1}+\cos ^{2} \frac{\theta}{2} \mathbf{a}_{2}
\end{array}
$$
where $ \mathbf{a}_1 $, $\mathbf{a}_2 $ are the Berry connections for the original Floquet states $|u_{1eq}\rangle $ and $|u_{2eq}\rangle $  .

Since the system is connected to a heat bath, the self energy takes the form $-i \Sigma^{<} / \Gamma=\left(\sigma_{z}+1\right) / 2$ and the occupation is given as as $f_{i}=\left(-i \Sigma_{i i}^{<}\right) / \Gamma$. 

The occupations are therefore, $f_{1}=(1+\cos \theta) / 2$ and $f_{2}=(1-\cos \theta) / 2.$

The original Hall conductivity $\sigma_{x y}^{0}$ for $E=0$ is given by setting $\theta=0 .$

We then obtain the photo-induced part of the Hall conductivity $\sigma_{x y}^{p} \equiv \sigma_{x y}-\sigma_{x y}^{0}$ as
$$
\begin{aligned}
\sigma_{x y}^{p} &=\frac{1}{4} \int d^{2} \mathbf{k} \sin ^{2} \theta\left[\nabla \times\left(\nabla \phi+\mathbf{a}_{2}-\mathbf{a}_{\mathbf{1}}\right)\right]_{z} \\
&=\int d^{2} \mathbf{k} \frac{A^{2}\left|\nu_{12}^{0}\right|^{2}}{4\left(d_{z}^{2}+A^{2}\left|\nu_{12}^{0}\right|^{2}\right)} \mathcal{F}
\end{aligned}
$$
where $\mathcal{F} \equiv\left(\nabla \times \mathbf{a}_{2}-\nabla \times \mathbf{a}_{\mathbf{1}}\right)_{z} .$ 

In the above integral we have used $\sin^2\theta = \frac{(d_x + i d_y)(d_x - id_y)}{|d|^2}$ and substitute $d_x, d_y$ and $d_z$ from $H_{F}$ to get $ \sin^2\theta = \frac{A^{2}\left|\nu_{12}^{0}\right|^{2}}{\left(d_{z}^{2}+A^{2}\left|\nu_{12}^{0}\right|^{2}\right)}$ 

If we assume that $A$ is sufficiently small, then using the approximation of the delta function from Cauchy's distribution, the Hall conductivity in the regime where the relaxation is much smaller than $A|\nu_{12}|$ is given by:
$$
\sigma_{x y}^{p}=\int d^{2} \mathbf{k} \frac{\pi E}{4 \Omega}\left|\nu_{12}^{0}\right| \delta\left(d_{z}\right) \mathcal{F}
$$
In the other regime where $A|\nu_{12}|$<<$\Gamma$, we can include the effect of relaxation by replacing the denominator with $4\left(d_{z}^{2}+A^{2}\left|\nu_{12}^{0}\right|^{2}+\Gamma^{2}\right),$ which leads to
$$
\sigma_{x y}^{p}=\int d^{2} \mathbf{k} \frac{\pi E^{2}}{4 \Gamma \Omega^{2}}\left|\nu_{12}^{0}\right|^{2} \delta\left(d_{z}\right) \mathcal{F}
$$

This photo-induced Hall conductivity is proportional to $E^2$ and describes the third-order nonlinear response. It is also proportional to the relaxation time 1/$\Gamma$, indicating that the nonlinear modulation arises from excited free electrons.

We therefore find that we have two regimes depending on the relative values of the parameter, which describes the crossover of $\sigma_{xy}^{p} \propto E$ and $\propto E^2/\Gamma$ in a manner similar to the shift current.

Since we have related the value of the photo-induced Hall response to topological invariants like the Berry curvature, it is now easy to see the affect of Time Reversal Symmetry on the same. 

The photoinduced Hall response is 0 when Time Reversal Symmetry is preserved. The contributions $\mathcal{F}$ at $\mathbf{k}$ and $\mathbf{-k}$ cancel each other as proved below:$^{[29]}$

The Berry curvature is defined as $$\mathcal{F}_{n}(\mathbf{k})=i \epsilon_{i j}\left\langle\partial_{ki} u_{n, k} \mid \partial_{kj} u_{n, k}\right\rangle =i \epsilon_{\mathrm{ij}} \int d \vec{r} \partial_{\mathrm{ki}} u_{n, k}(\vec{r})^{*} \partial_{\mathrm{kj}} u_{n, k}(\vec{r})$$ where $\epsilon_{i j}$ is the Levi civita symbol.

Under time-reversal, the sign of momentum flips and the new $\mathcal{F}_{n}(k)$ is defined as
$$ \mathcal{T} \mathcal{F}_{n}(k)=i\epsilon_{\mathrm{ij}}\left\langle\partial_{\mathrm{ki}} \mathcal{T} \mathrm{u}_{n, k} \mid \partial_{\mathrm{kj}} \mathcal{T}\mathrm{u}_{n, k}\right\rangle$$
$$
=i \epsilon_{\mathrm{ij}} \int d \vec{r} \partial_{\mathrm{ki}} \mathcal{T}u_{n, k}(\vec{r})^{*} \partial_{\mathrm{kj}} \mathcal{T}u_{n, k}(\vec{r})
$$
$$
=i \epsilon_{\mathrm{ij}} \int d \vec{r} \partial_{\mathrm{ki}} u_{n,-k}(\vec{r}) \partial_{\mathrm{kj}} u_{n,-k}(\vec{r})^{*}=i \epsilon_{\mathrm{ij}} \int d \vec{r} \partial_{\mathrm{kj}} u_{n,-k}(\vec{r})^{*} \partial_{\mathrm{ki}} u_{n,-k}(\vec{r})
$$
Swapping the dummy indices $i$ and $j$ here
$$
\mathcal{T F}_{n}(k)=i \epsilon_{\mathrm{ji}} \int d \vec{r} \partial_{\mathrm{ki}} u_{n,-k}(\vec{r})^{*} \partial_{\mathrm{kj}} u_{n,-k}(\vec{r})=-i \epsilon_{\mathrm{ij}} \int d \vec{r} \partial_{\mathrm{ki}} u_{n,-k}(\vec{r})^{*} \partial_{\mathrm{kj}} u_{n,-k}(\vec{r})=-\mathcal{F}_{n}(-k)
$$
If the system has time-reversal symmetry
$$
\mathcal{T F}_{n}(k)=\mathcal{F}_{n}(k)
$$
So we have
$$
\mathcal{F}_{n}(-k)=-\mathcal{F}_{n}(k)
$$
For systems with time-reversal symmetry, $\mathcal{F}_{n}(k)$ is an odd function of $\mathrm{k}$. When we integrate over the Brillouin zone, it evaluates to 0.

When time-reversal symmetry is broken, there is a finite photo-induced Hall response which acts as a correction to $\sigma_{xy}.$ The contribution $\mathcal{F}$ is equal to the difference between the Berry curvatures of the valence and conduction bands respectively.

This effect can be extended to describe non-linear Kerr rotations when the probe electric field has a non-zero frequency instead of being a static field. The application of a magnetic field leads to rotation of polarization plane which is known as nonlinear Kerr rotation.$^{[30]}$

The effects may also be finite even in the case of special T-symmetric
cases. The Berry curvature is related to the phase picked up by a state as the Hamiltonian is adiabatically varied in a parameter space. So far, our parameter space has been k-space, but this can be generalised. If the Hamiltonian depends on some other parameter such as Q, we can still have a non-trivial photo-induced Hall effect.


###<center>2.5 Inversion Symmetry in 1D Chains</center>

We use a 1-D model consisting of an infinite chain of two sub-lattices A and B. This model, with bond strength alternation, can be applied to a wide variety of organic conductors $^{[36]}$. The Hamiltonian is as follows:

$\sum_{i}^{} \frac{1 + Q_2 (-1)^i}{2} (e^{iFt} {c_i}^\dagger c_{i+1} + h.c.) + Q_1 (-1)^i {c_i}^\dagger c_i + {Q'}_1 (-1)^i (e^{2iFt} {c_i}^\dagger c_{i+2} + h.c.)  $

The first term represents the interaction between atoms present on nearest-neighbor sites. The first term has the $Q_2 (-1)^i$ expression to accomodate the different bond lengths (and hence strengths) of A-B and B-A links. The second term describes the self energy of site occupation. Since the number of total electrons remains constant, a term of the form $\sum_i {c_i}^\dagger c_i = N$ has been omitted. The last term is due to the next-nearest-neighbor interaction. This has been arbitratily added to break T symmetry, and does not have any physical basis. 

The hamiltonian may be fourier transformed using $ c_i = \frac{\sqrt{a}}{\sqrt{ 2 \pi}} \int_{BZ} dk c_k e^{ikx}$ to get the following 2-by-2 hamiltonian (as may be expected by the two degrees of freedom per unit cell):

$H_0 = \text{cos} \frac{k}{2} \sigma_x + Q_2 sin \frac{k}{2} \sigma_y + (Q_1 + {Q'}_1 \text{sin}k)\sigma_z$

Further calculations become easier once we expand this into a $2 \times 2$ form:

\begin{equation}
  H_0 (k) = 
  \begin{bmatrix}
   Q_1 + {Q'}_1 \text{sin}k & \text{cos} \frac{k}{2} - i Q_2 \text{sin} \frac{k}{2} \\
   \text{cos} \frac{k}{2} + i Q_2 \text{sin} \frac{k}{2} & Q_1 + {Q'}_1 \text{sin}k \\
  \end{bmatrix}
\end{equation}

We integrate $H_0 (k + i A e^{-i \Omega t} - i A^* e^{i \Omega t})$ with respect to t to arrive at the Floquet Hamiltonian $H_{mn}$:

\begin{equation}
  H_{mn} = 
  \frac{1}{T} \int_0^T dt e^{i (m - n) \Omega t} 
    \begin{bmatrix}
      Q_1 + {Q'}_1 \text{sin}k' & \text{cos} \frac{k'}{2} - i Q_2 \text{sin} \frac{k'}{2} \\
   \text{cos} \frac{k'}{2} + i Q_2 \text{sin} \frac{k'}{2} & Q_1 + {Q'}_1 \text{sin}k' \\
   \end{bmatrix}
\end{equation}

Here $k' = k + i A e^{-i \Omega t} - i A^* e^{i \Omega t} $.

Integrating, this transforms into a constant term plus a bessel function term:


\begin{equation}
H_{m n}(k)=\left(\begin{array}{cc}
Q_{1}-n \Omega & 0 \\
0 & -Q_{1}-n \Omega
\end{array}\right) \delta_{m n}+\left(\begin{array}{cc}
C_{m n} & A_{m n} \\
B_{m n} & -C_{m n}
\end{array}\right)
\end{equation}

Here $A_{m n}=\left(t+Q_{2} / 2\right) e^{-i k / 2} J_{m-n}(-F / 2)+\left(t-Q_{2} / 2\right) e^{i k / 2} J_{m-n}(F / 2), B_{m n}=$
$\left(t+Q_{2} / 2\right) e^{i k / 2} J_{m-n}(F / 2)+\left(t-Q_{2} / 2\right) e^{-i k / 2} J_{m-n}(-F / 2),$ and $C_{m n}=Q_{j}^{\prime} J_{m-n}$
$(F)(-i)^{m-n}\left[(-1)^{m-n} e^{i k}-e^{-i k}\right] /(2 i),$ where $F=e E a / \Omega$ with lattice spacing
$a$, and $J_{n}(x)$ is the $n$ th Bessel function.

We may also numerically integrate the above matrix equation to obtain the Floquet Hamiltonian $H_{mn}$. 
We show this for the component $({H_{00}})_{11}$ below:

In arbitrary units, set $Q_1 = 1$, ${Q_1}' = 2$, $\Omega = 3$, $a = 4$.  The two independent variables then become $k$ and $E$, and we will compare the dependence of $({H_{00}})_{11}$ on these variables. 

We get the following integral for $({H_{00}})_{11}$ :

\begin{equation}
  ({H_{00}})_{11} = \frac{1}{T} \int_0^T (Q_1 + Q'_1 \sin(k + iAe^{-i \Omega t} - iA*e^{i \Omega t})dt 
\end{equation}

Integrating out the constant term:

\begin{equation}
  ({H_{00}})_{11} = Q_1 + Q'_1 {\int_0}^T \sin(k + iAe^{-i \Omega t} - iA*e^{i \Omega t})dt
\end{equation}

We can now analyse numerically the second term of this equation (since the other is trivial):

\begin{equation}
{\int_0}^T \sin(k + iAe^{-i \Omega t} - iA*e^{i \Omega t})dt
\end{equation}

Now we analyse this term as a function of the electric field $E$, keeping all other variables fixed.












```
from sympy import *
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
from scipy.integrate import quad

def integrand(t,k1,e1,omega1):
  val = math.sin(k1 + 2*(e1/omega1)*sin(omega1*t))
  return val

omega = 1
k = 1

def integral(k2,e2,omega2):
  """This function calculates the integral in equation () given a value of k, omega and e"""
  val= quad(integrand,0,(2*math.pi)/omega2,args=(k2,e2,omega2))
  return val[0]

x = np.linspace(0,10,1000)
integralx =np.zeros(len(x))
i=0

for num in x:
  integralx[i] = integral(k,num,omega)
  i=i+1

plt.plot(x,integralx)
```

It is instructive to compare the result of this numerical integration to the analytically integrated solution in the paper. The $(H_{00})_{11}$ value is found to be:

\begin{equation}
\begin{split}
(H_{00})_{11} &= Q_1 + C_{00} \\
&= Q_1 + Q'_1 J_0 (\frac{eEa}{\Omega}) \times \sin k
\end{split}
\end{equation}

What we have graphed is then the term $J_0 (\frac{eEa}{\Omega}) \times \sin k$ ( $J_0$ here is the 0th bessel function of the first kind.) The graph shows its dependence on E, while all other variables are constant. The bessel function dependence is quite clearly visible (a graph of the bessel function of the first kind may be found at https://en.wikipedia.org/wiki/Bessel_function).

In this manner, any component of $H_{mn}$ may be obtained by numerical integration, and hence the full Floquet matrix may be found.

Now we plot the 2-D vector field $\tilde{A}(k,Q_2) = |\nu^0_{12}|^2 (R_k,R_{Q_2})$ in the $(k,Q_2)$ plane. This quantity is directly related to the shift vectors $R_k$ and $R_{Q_2}$ and can be used to make instructive symmetry based inferences from the following plots.

One may also perform the following calculation:

\begin{equation}
\begin{split}
 \nabla \times \tilde{A} &= |\nu^0_{12}|^2 (\partial_k R_{Q_2} - \partial_{Q_2} R_k)\\
 &= |\nu^0_{12}|^2 (\partial_k \partial_{Q_2} \phi_{12} - \partial_{Q_2} \partial_k \phi_{12} +  \nabla \times a_1 - \nabla \times a_2)\\
 &= |\nu^0_{12}|^2 (\mathcal{F}_1 - \mathcal{F}_2)\\
 &\equiv |\nu^0_{12}|^2 (\mathcal{F})
\end{split}
\end{equation}

Thus the curl of the field $\tilde{A}$ is thus related to the difference of the Berry curvatures of the valence and conduction bands.  

The calculation of $\tilde{A}$ warrants the calculation of some prelimnaries:

1. The eigenvectors $u_i$ of $H_0$
2. The berry connection $-i \langle{u_i}|{\partial_k u_i}\rangle$
3. The matrix $\nu^0$, defined as $\frac{\partial H_0}{\partial k}$

The following python code calculates the eigenvectors $u_i$ (for given $Q_1$, $Q'_1$ and $Q_2$):




```
from sympy import *
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
from scipy.integrate import quad

k = Symbol("k",real=True)
Q1 = 0
Q1dash = 1
Q2 = 1

H0 = Matrix([[Q1+Q1dash*sin(k),cos(k/2)-I*Q2*sin(k/2)],[cos(k/2)+I*Q2*sin(k/2),Q1 + Q1dash*sin(k)]])
returnmatrix = [H0.eigenvects()[0][2],H0.eigenvects()[1][2]]
returnmatrix

```

In the above code we've taken $Q_1 = 0$, $Q'_1 = 1$, $Q_2 =1$, and the eigenvectors obtained are 

\begin{equation}
u_1 = 
\begin{bmatrix}
-e^{\frac{-ik}{2}} \\
1 \\
\end{bmatrix}
\text{and }
u_2 = 
\begin{bmatrix}
e^{\frac{-ik}{2}} \\
1 \\
\end{bmatrix}
\end{equation}

In this manner we may compute $u_1$ and $u_2$ for any given values of the parameters. In the code below we use these to compute the berry connection, $\nu_0$ and $\tilde{A}$


```

import math
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
from scipy import linalg

Q1 = 1
Q1dash = 0

def u(Q1,Q1dash,Q2,k): 
  #This function calculates the eigenvectors of the 2x2 Hamiltonian, given parameters
  H0 = np.array([[Q1+Q1dash*math.sin(k),math.cos(k/2)-1j*Q2*math.sin(k/2)],[math.cos(k/2)+1j*Q2*math.sin(k/2),-Q1 - Q1dash*math.sin(k)]])
  val, eigen = linalg.eig(H0)
  returnmatrix = eigen
  return returnmatrix


def ak(Q1,Q1dash,Q2,k):
  #This function calculates the berry curvatures by calling the eigenvector
  #function and then differentiating
  u1 = u(Q1,Q1dash,Q2,k)[:,0]
  du1dk = (u(Q1,Q1dash,Q2,k+0.001)[:,0] - u(Q1,Q1dash,Q2,k-0.001)[:,0])/0.002
  u2 = u(Q1,Q1dash,Q2,k)[:,1]
  du2dk = (u(Q1,Q1dash,Q2,k+0.001)[:,1] - u(Q1,Q1dash,Q2,k-0.001)[:,1])/0.002
  a1 = -1j*np.vdot(u1,du1dk)
  a2 = -1j*np.vdot(u2,du2dk)
  a = np.real(a1 - a2)
  return a

def aQ2(Q1,Q1dash,Q2,k):
  #This function calculates the berry curvatures by calling the eigenvector
  #function and then differentiating
  u1 = u(Q1,Q1dash,Q2,k)[:,0]
  du1dk = (u(Q1,Q1dash,Q2+0.001,k)[:,0] - u(Q1,Q1dash,Q2-0.001,k)[:,0])/0.002
  u2 = u(Q1,Q1dash,Q2,k)[:,1]
  du2dk = (u(Q1,Q1dash,Q2+0.001,k)[:,1] - u(Q1,Q1dash,Q2-0.001,k)[:,1])/0.002
  a1 = -1j*np.vdot(u1,du1dk)
  a2 = -1j*np.vdot(u2,du2dk)
  a = np.real(a1 - a2)
  return a


def dphidk(Q2,k):
  #This part calculates the derivative of the phase of v_0^12 WRT k
  temp = -(math.atan2(Q2*math.cos(k/2),math.sin(k/2)) - math.atan2(Q2*math.cos((k+0.001)/2),math.sin((k+0.001)/2)))*1000
  return temp

def dphidQ2(Q2,k):
  #This part calculates the derivative of the phase of v_0^12 WRT Q2
  temp = -(math.atan2(Q2*math.cos(k/2),math.sin(k/2)) - math.atan2((Q2+0.001)*math.cos(k/2),math.sin(k/2)))*1000
  return temp


def Rk(Q1,Q1dash,Q2,k):
  #This uses previously defined functions to derive the shift vector Rk
  return (dphidk(Q2,k) + ak(Q1,Q1dash,Q2,k))

def RQ2(Q1,Q1dash,Q2,k):
  #This uses previously defined functions to derive the shift vector Rk
  return (dphidQ2(Q2,k) + ak(Q1,Q1dash,Q2,k))

def v012sq(Q2,k):
  #This calculates the square of v_0^12
  return (math.sin(k/2))**2 + (Q2**2)*((math.cos(k/2))**2)

#Create arrays of coordinates:
karr = np.linspace(-1.5*math.pi,1.5*math.pi,20)
Q2arr = np.linspace(-0.5,0.5,15)

#Create arrays of vectors:
Rkarr = np.zeros([15,20])
RQ2arr = np.zeros([15,20])

#Fill arrays of vectors with appropriate values:

for i in range(len(karr)):
  for j in range(len(Q2arr)):
    Rkarr[j,i] = v012sq(Q2arr[j],karr[i])*Rk(Q1,Q1dash,Q2arr[j],karr[i])

for i in range(len(karr)):
  for j in range(len(Q2arr)):
    RQ2arr[j,i] = 0.3*v012sq(Q2arr[j],karr[i])*RQ2(Q1,Q1dash,Q2arr[j],karr[i])

C = np.hypot(Rkarr,RQ2arr)
#plot:
plt.quiver(karr,Q2arr,Rkarr,RQ2arr,C)
```

Let us look at the first plot and make inferences based on the same. The sum of $A_k$ at $k$ and $-k$ satisfying the condition $E_2(k) - E_1(k) = \hbar \omega$ (blue dots in figure below) corresponds to the shift current $J$ proportional to $E^2$. When $Q_2$ is non-zero, that is, when parity symmetry is broken, this sum does not vanish. This can be seen by adding up $x$-components of the vector $\tilde{A}$ at $k$ and $-k$.

The corresponding quantity for $Q_2$, that is $\tilde{A}_{Q_2}$, gives us the change in the bond dimerization $B=\Sigma_{i}(-1)^{i}\left(c_{i}^{\dagger} c_{i+1}+ \text{h.c.}\right)$ $^{[37]}$. Changing $Q_2$ in the Hamiltonian corresponds to changing the difference between the (A-B) and (B-A) bond strengths. Since $Q_2$ does not change under time reversal, the $\tilde{A}_{Q_2}$ ($y$) components at $k$ and $-k$ add up to zero. 

From the previous calculation we know that $\nabla \times  \tilde{A} = |\nu^0_{12}|^2 (\mathcal{F}_1 - \mathcal{F}_2) $. It is a simple exercise to show that the integral of $\mathcal{F_i}$ over the first Brilloin zone gives us the expectation value of position, and hence the polarization. The integral of $\tilde{A}$, then, will give us the difference of the polarization between the valence and conduction bands. 

The value of $\tilde{\mathcal{F}}$ is related to the change in the bond dimerization $B$, which is a third order response proportional to $E^3$. This can be obtained  by replacing $k_y$ by $Q_2$ in the equation for the Hall conductivity from Roshni's section, since now, instead of $k_y$ and $k_x$, we use $Q_2$ and $k_x$ as varying parameters in the Hamiltonian. 

Also note that there is an antivortex for $\tilde{A}$ at $(k,Q_2) = (\pm \pi, 0)$, corresponding to a maxima of the Berry curvature. The vortex at $(0,0)$ corresponds to a singularity at $\nu^0_{12}$ vanishes. 

The graph for $Q_1 =1 $ and  $Q'_1 = 0.1$ is shown below. Note that this configuration breaks time reversal symmetry.

```

import math
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
from scipy import linalg

Q1 = 1
Q1dash = 0.4

def u(Q1,Q1dash,Q2,k):
  #This function calculates the eigenvectors of the 2x2 Hamiltonian, given parameters
  H0 = np.array([[Q1+Q1dash*math.sin(k),math.cos(k/2)-1j*Q2*math.sin(k/2)],[math.cos(k/2)+1j*Q2*math.sin(k/2),-Q1 - Q1dash*math.sin(k)]])
  val, eigen = linalg.eig(H0)
  returnmatrix = eigen
  return returnmatrix


def ak(Q1,Q1dash,Q2,k):
  #This function calculates the berry curvatures by calling the eigenvector
  #function and then differentiating
  u1 = u(Q1,Q1dash,Q2,k)[:,0]
  du1dk = (u(Q1,Q1dash,Q2,k+0.001)[:,0] - u(Q1,Q1dash,Q2,k-0.001)[:,0])/0.002
  u2 = u(Q1,Q1dash,Q2,k)[:,1]
  du2dk = (u(Q1,Q1dash,Q2,k+0.001)[:,1] - u(Q1,Q1dash,Q2,k-0.001)[:,1])/0.002
  a1 = -1j*np.vdot(u1,du1dk)
  a2 = -1j*np.vdot(u2,du2dk)
  a = np.real(a1 - a2)
  return a

def aQ2(Q1,Q1dash,Q2,k):
  #This function calculates the berry curvatures by calling the eigenvector
  #function and then differentiating
  u1 = u(Q1,Q1dash,Q2,k)[:,0]
  du1dk = (u(Q1,Q1dash,Q2+0.001,k)[:,0] - u(Q1,Q1dash,Q2-0.001,k)[:,0])/0.002
  u2 = u(Q1,Q1dash,Q2,k)[:,1]
  du2dk = (u(Q1,Q1dash,Q2+0.001,k)[:,1] - u(Q1,Q1dash,Q2-0.001,k)[:,1])/0.002
  a1 = -1j*np.vdot(u1,du1dk)
  a2 = -1j*np.vdot(u2,du2dk)
  a = np.real(a1 - a2)
  return a


def dphidk(Q2,k):
  #This part calculates the derivative of the phase of v_0^12 WRT k
  temp = -(math.atan2(Q2*math.cos(k/2),math.sin(k/2)) - math.atan2(Q2*math.cos((k+0.001)/2),math.sin((k+0.001)/2)))*1000
  return temp

def dphidQ2(Q2,k):
  #This part calculates the derivative of the phase of v_0^12 WRT Q2
  temp = -(math.atan2(Q2*math.cos(k/2),math.sin(k/2)) - math.atan2((Q2+0.001)*math.cos(k/2),math.sin(k/2)))*1000
  return temp


def Rk(Q1,Q1dash,Q2,k):
  #This uses previously defined functions to derive the shift vector Rk
  return (dphidk(Q2,k) + ak(Q1,Q1dash,Q2,k))

def RQ2(Q1,Q1dash,Q2,k):
  #This uses previously defined functions to derive the shift vector Rk
  return (dphidQ2(Q2,k) + aQ2(Q1,Q1dash,Q2,k))

def v012sq(Q2,k):
  #This calculates the square of v_0^12    
  return (math.sin(k/2))**2 + (Q2**2)*((math.cos(k/2))**2)

#Create arrays of coordinates:
karr = np.linspace(-1.5*math.pi,1.5*math.pi,20)
Q2arr = np.linspace(-0.5,0.5,15)

#Create arrays of vectors:
Rkarr = np.zeros([15,20])
RQ2arr = np.zeros([15,20])

#Fill arrays of vectors with appropriate values:

for i in range(len(karr)):
  for j in range(len(Q2arr)):
    Rkarr[j,i] = v012sq(Q2arr[j],karr[i])*Rk(Q1,Q1dash,Q2arr[j],karr[i])

for i in range(len(karr)):
  for j in range(len(Q2arr)):
    RQ2arr[j,i] = 0.3*v012sq(Q2arr[j],karr[i])*RQ2(Q1,Q1dash,Q2arr[j],karr[i])

C = np.hypot(Rkarr,RQ2arr)
#plot:
plt.quiver(karr,Q2arr,Rkarr,RQ2arr,C)
```

####<center> Solution:

The instantaneous width of the well is $w(t)=w_{0}+\left(v_{2}-v_{1}\right) t$ and the rate of increase of this width is $\Delta v=v_{2}-v_{1} .$ 

The wavefunction $\psi(x, t)$ must satisfy Schrodinger's equation
$$
i \hbar \frac{\partial \psi}{\partial t}=-\frac{\hbar^{2}}{2 m} \frac{\partial^{2} \psi}{\partial x^{2}}
$$
subject to the boundary conditions $\psi\left(w_{2}(t), t\right)=\psi\left(w_{1}(t), t\right)=0 .$ 

(a) The wave function in the adiabatic approximation is given by:

$$
\Psi_{n}(x, t)=\sqrt{\frac{2}{w(t)}} \sin \left(\frac{n \pi\left(x-w_{1}(t)\right)}{w(t)}\right) \exp \left(i\left(-\int_{0}^{t} \frac{E_{n}\left(t^{\prime}\right)}{\hbar} d t^{\prime}+\gamma_{n}(t)\right)\right)
$$

where $w(t) = w_1(t)-w_2(t).$

(b) The geometric phase may be calculated by
$$
\gamma_{n}(t)=i \int_{R(0)}^{R(t)}\left\langle n\left(R\left(t^{\prime}\right)\right)\left|\nabla_{R}\right| n\left(R\left(t^{\prime}\right)\right)\right\rangle d R^{\prime}
$$
where $R(t)=\left(w_{1}(t), w_{2}(t)\right)$ and $|n(R(t))\rangle$ is the $n$ th instantaneous eigenmode with lower and upper walls $w_{1}(t)$ and $w_{2}(t)$ respectively:
$$
|n(R)\rangle=\sqrt{\frac{2}{\left(w_{2}-w_{1}\right)}} \sin \left(\frac{n \pi\left(x-w_{1}\right)}{w_{2}-w_{1}}\right)
$$
(c) Calculating the Berry connection:

For the sake of abbreviation, let $u=n \pi\left(x-w_{1}\right) /\left(w_{2}-w_{1}\right)$ and $w=w_{2}-w_{1} .$ Then
$$
\begin{aligned}
\left\langle n(R)\left|\frac{\partial}{\partial w_{1}}\right| n(R)\right\rangle &=\frac{2}{w} \int_{w_{1}}^{w_{2}} \sin (u)\left(\frac{\sin (u)}{2 w}+(\frac{u}{w} -\frac{n\pi}{w})\cos (u)\right) d x \\
&=\frac{1}{w} \int_{0}^{n \pi} \sin ^{2}(u)+2 u \sin (u) \cos (u) - (n\pi)sin(2u)d u 
\end{aligned}
$$


This trigonometric integral may be evaluated directly, or by observing that $d\left(u \sin ^{2}(u)\right) / d u$ is exactly the integrand. Either way, the integral vanishes. By symmetry, $\left\langle n(R)\left|\partial / \partial w_{2}\right| n(R)\right\rangle$ must vanish also. Thus, $\left\langle n\left(R\left(t^{\prime}\right)\right)\left|\nabla_{R}\right| n\left(R\left(t^{\prime}\right)\right)\right\rangle=0$ and the geometric phase $\gamma_{n}(t)$ is trivial.

Additional:

Note that the exact set of solutions, without the adiabatic approximation is:
$$
\psi_{n}(x, t)=\sqrt{\frac{2}{w(t)}} \sin \left(\frac{n \pi\left(x-w_{1}(t)\right)}{w(t)}\right) \exp \left(i \theta(x, t)-i \frac{m E_{n}^{0} w_{0} t}{\hbar w(t)}\right)
$$
where $E_{n}^{0}=\left(n^{2} \hbar^{2} \pi^{2}\right) / 2 m w_{0}^{2}$ is the energy eigenvalue for the $n$ -th eigenstate infinite potential well with constant width $w_{0}$ and
$$
\theta(x, t)=\frac{m\left(\Delta v x^{2}+2 v_{1} w_{0} x-v_{1}^{2} w_{0} t\right)}{2 \hbar w(t)}
$$
Note that $\theta(x, t)$ is independent of the mode number $n$. The solutions are very similar to the adiabatic solutions.

### <center>4.1 Gaussian Integration</center>
For a moment, we digress to Gaussian integration, and show that the correlation function indeed equals the Green's function [33], for completeness.
It forms an important tool for path integration, the technique used to derive the Keldysh formalism. 

Consider a discretised path integral of the form $$\mathcal Z=\int_{-\infty}^\infty \prod_i d\phi_i e^{-\phi^TG^{-1}\phi +h^T\phi},$$ where $h$ and the field $\phi$ are real column vectors, and the operator $G$ is a real symmetric matrix. 
Diagonalise $G$ with $U$ such that $\tilde G=UGU^{-1}$, and observe that

 $$(U\phi-\tilde G Uh)^T\tilde G^{-1}(U\phi-\tilde G Uh)-h^TU^{-1}\tilde G Uh=(\phi^TU^{-1}-h^TU^{-1}\tilde G)(\tilde G^{-1}U\phi-Uh)-h^TU^{-1}\tilde G Uh=\phi^T U^{-1}\tilde G^{-1}U\phi+h^TU^{-1}\tilde GUh-h^TU^{-1}U\phi-\phi^T U^{-1}Uh-h^TU^{-1}\tilde G Uh=\phi^TG^{-1}\phi-h^T\phi-\phi^T h=2(\phi^TG^{-1}\phi/2-h^T\phi).$$ 

Therefore, write, using the extreme left and right hand sides, $$(\chi^T \tilde G^{-1} \chi-h^TU^{-1}\tilde G Uh)/2=(\chi^T \tilde G^{-1} \chi-h^TGh)/2=\phi^TG^{-1}\phi/2-h^T\phi.$$

Change variables in the path integral, $\phi_i \to \chi_i$. The transformation is unitary, so the Jacobian is simply 1.

 Now, $$\mathcal Z=\int_{-\infty}^\infty \prod_i d\chi_i e^{-(\chi^T \tilde G^{-1} \chi-h^TGh)/2}=e^{h^TGh/2}\int_{-\infty}^\infty \prod_i d\chi_i e^{-\chi^T \tilde G^{-1} \chi}.$$
  The integral factorises, since $\tilde G^{-1}$ is diagonal. Thus, our problem reduces to the solution of a typical single variable Gaussian integral, one for each eigenvalue of $\tilde G$. The product of all its eigenvalues appears as $\det(G)$. In particular, the final answer is simply $$\sqrt{\det(2\pi G)}e^{h^TGh/2}.$$

The characteristic function, $$\langle{e^{k^T\phi}}\rangle=\int_{-\infty}^\infty \prod_i d\phi_i e^{-\phi^TG^{-1}\phi +(h+k)^T\phi}.$$
 Note that $$(\partial_{k_i}\partial_{k_j})_{k=0}\ln\langle{e^{k^T\phi}}\rangle=\langle{\phi_i\phi_j}\rangle.$$ In this case, the $k$-dependent part of $$\ln\langle{e^{k^T\phi}}\rangle$$ is that of $$(h+k)^TG(h+k)/2,$$ i.e. $h^TGk+k^TGk/2$. Finally, $$(\partial_{k_i}\partial_{k_j})_{k=0} (h^TGk+k^TGk/2)=G_{ij}.$$
The continuous version of the path integral will often be expressible in the form $$\mathcal Z\sim \int[d\phi] e^{-\phi(x)\delta(x-x')\mathcal D(x,x')\phi(x')},$$ with $\mathcal D$ being some linear differential operator. 

From the form of this integral, $\delta(x-x')=\mathcal D(x,x') G(x,x')$ will yield a Green's function, 
the inverse of $\delta(x-x')\mathcal D(x,x')$, and also the correlator $\langle{x,x'}\rangle$. 
For a free boson, we find the 2-point correlation function $G(x,x')=\langle{\phi(x)\phi(x')}\rangle$ to elucidate the discussion. Moreover, $$S_E=g\int \text{d}^2x[\partial_\mu \phi\partial^\mu\phi+m^2\phi^2]/2.$$ 
We need to rewrite this to identify the correlator. 

Integrate the first term by parts to shift both the derivatives to a single term. Also throw in a Dirac's delta and change appropriate instances $x$ to $x'$ so that we have $$S_E=g\int \text{d}^2x \text{d}^2x'\delta(x-x')[-\phi(x)\partial^2\phi(x')+m^2\phi(x)\phi(x')]/2=\int \text{d}^2x \text{d}^2x'\phi(x) A(x,x')\phi(x')/2,$$ with $A(x,x')=g\delta(x-x') (-\partial^2+m^2)/2$. Hence identify, from the continuous generalisation discussed, $$G=A^{-1}\implies g(-\partial_x^2+m^2)G(x,x')=\delta(x-x').$$

 Noting that rotational and translational invariance enforces $G(x,x')\equiv G(|x-x'|)=G(r)$, integrate both sides of the equation as $x':0\to r$. 
 
 For $m=0$, we get a single equation, $G'(r)=-1/(2\pi g r)$, yielding $G(r)=-\ln(r)/(2\pi g)$. For non-zero mass, we have the $0^{th}$ order modified Bessel equation, $G''+G'/r-m^2G=0$, the only physically valid (vanishing at infinity) solution of which is in terms of $K_0$, the $0^{th}$ order modified Bessel function, $G(r)=K_0(mr)/(2\pi g)\overset{mr\gg 1}{\sim} e^{-mr}$: Correlation lengths for massive bosonic fields decay exponentially for large distances [32].

The calculation for a free fermion is an important step in the derivation of the Keldysh formalism.

### <center>4.2 Demonstration Code to help visualise discussed phenomena in action</center>

This section numerically solves the Schrodinger equation for a user-derined potential in one dimension and analyses the solution [25]. In the process, it teaches a cute technique for visualising kets as vectors.
We'll take a general potential $V(x)$, and specify the form only in the code. It may easily be modified by the user. The Schrodinger's equation then reads $$\frac{-\hbar^2}{2m}\frac{\partial^2\psi}{\partial x^2}+V(x)\psi=E\psi.$$
We take $\hbar=m=1$ for simplicity; we can easily put these back in at the end. We work with units of length where $V(x+2\pi)=V(x)$ when we assume periodic $V(x)$. This section will help us appreciate Bloch's theorem, and assist the readers help themselves investigate any curiosity they might have.

We work in a long box of length $100$ units. We must first discretise all our quantities. We do this with a spacing of $h=0.1$ units. Sample $V(x)$ on an array $[0,0.1,\dots,500]$ to get a row vector; write this row vector as a diagonal matrix $\mathbf{V}$, in essence, sampling the 2-variable function $V(x)\delta(x-x')\to V(i)\delta_{ij}$. Meanwhile, sample the unknown $\psi(x)$ on these points to get an unknown column vector $\mathbf{\Psi}$. The differentiation operator is discretised so that it yields a finite difference forumula; call it $\mathbf{D}$. Then, we have an eigenvalue equation
$$(-\frac{\mathbf{D}^2}{2}+\mathbf{V})\mathbf{\Psi}=E\mathbf{\Psi}.$$


```
import numpy as np
import matplotlib.pyplot as plt
L=100
h=0.1 
dim=int(L/h) #To avoid complexity we've written a simplistic code; try to ensure that L/h is an integer
grid=np.arange(0,L,h)

def potential(x):
  return np.sin(x)-1 #User defined potential goes here
  #return np.heaviside(x-3*L/4,0.5)+np.heaviside(L/4-x,0.5)

temppot=potential(grid) #Sampling the potential into a row
plt.plot(grid,temppot) #Plotting the potentiak
plt.xlabel("x")
plt.ylabel("V(x)")
#plt.xlim(0,30)

V=np.zeros((dim,dim))
for i in range(0,dim):
  V[i,i]=temppot[i] #Converting it into matrix form

D2=np.zeros((dim,dim))
for i in range(0,dim-1):
  D2[i,i+1]=1
  D2[i,i]=-2
for i in range(1,dim):
  D2[i,i-1]=1
  D2[i,i]=-2
D2=D2/h**2 #Finishes defining the double differentiation matrix
H=-D2/2+V #Defines Hamiltonian
E,Psi=np.linalg.eigh(H)
idx = E.argsort()[::-1] #Sorting by energy
E=E[idx]
Psi=Psi[:,idx]

#Analysing Output

for i in range(990,900,-5):
  plt.figure()
  plt.plot(grid,Psi[:,i])
  plt.xlabel("x")
  plt.ylabel("Psi(x), E="+str(E[i]))

plt.figure()
plt.plot(E,'o')
plt.xlabel("Index")
plt.ylabel("Eigen-energy")
#plt.xlim(960,1000)
#plt.ylim(0,2)
plt.xlim(900,1000)
plt.ylim(-2,2)
```

###<center> Illustrative code to calculate the Berry connection for the eigenstates of a general 2X2 Hamiltonian.


The following code simply takes as input a 2X2 matrix, and is capable of returning the values of $d_z$, $d_y$ and $d_x$ and $\theta$ and $\phi$ if we write the Hamiltonian in standard form as $\epsilon \mathbb{I}$ + $d \cdot \sigma$. Note that to define the Berry connection we need to know the functional form connecting $\phi$ and $k$. \begin{bmatrix}
0&exp(-ika)\\
exp(ika)&0
\end{bmatrix}

### <center>4.4 Tight binding method; Dirac fermion operators</center>

The tight binding method $^{[38]}$ aims to find the band structure of solids by using a superposition of a set of wavefunctions $\psi_i$ localised to each atom $i$. We start with a vacuum state (that is no electron on any atom) $|{0}\rangle$ and use dirac creation (annihilation) operators to create (annihilate) an electron on a particular atom.

We call such a creation (annihilation) operator that creates (annihilates) an electron on atom $i$ as ${c^\dagger}_i$ ($c_i$). Note that the creation and annihilation operators are hermitian adjoints of each other. In particular, the action of the creation operator $c^\dagger_i$ on the vacuum state is as follows:

\begin{equation}
    c^\dagger_i |{0}\rangle = |{\psi_i}\rangle
\end{equation}

Here $|{\psi_i}\rangle$ denotes the wave function of a single electron localised on atom $i$. 

When there is more than one electron, it is necessary to anti-symmetrise the wavefunction since the electrons are fermions. This is done with the antisymmeterisation operator $\mathcal{A}$ as follows:

\begin{equation}
    \mathcal{A} |{\psi_i}\rangle |{\psi_j}\rangle = \frac{1}{2!} \left(  |{\psi_i}\rangle |{\psi_j}\rangle - |{\psi_j}\rangle |{\psi_i}\rangle \right) 
\end{equation}

The action of the antisymmetrisation operator can be analogously defined for states with more than two electrons by means of the Slater determinant (we assume the reader knows this from CH107). 

Antisymmeterisation is implicitly assumed upon action of the creation and annihilation operators. This allows us to define a state solely on the basis of which orbitals are occupied and which are not, i.e. the <i>occupation number formalism</i>. For example, for a system with 5 atoms, the vacuum state will be:

\begin{equation}
    |{0}\rangle = |{0,0,0,0,0}\rangle
\end{equation}

Upon action of the creation operator $c^\dagger_3$:

\begin{equation}
    c^\dagger_3 |{0,0,0,0,0}\rangle = |{0,0,1,0,0}\rangle = |{\psi_3}\rangle
\end{equation}

Upon action of another creation operator $c^\dagger_5$:

\begin{equation}
    c^\dagger_5 |{0,0,1,0,0}\rangle = |{0,0,1,0,1}\rangle = \frac{1}{2!} \left(  |{\psi_3}\rangle |{\psi_5}\rangle - |{\psi_5}\rangle |{\psi_3}\rangle \right) 
\end{equation}

Annihilation:

\begin{equation}
    c_5 |{0,0,1,0,1}\rangle = |{0,0,1,0,0}\rangle = |{\psi_3}\rangle
\end{equation}

At this point it is instructive to note the action of the annihilation $c_i$ operator on a state with orbital $i$ unoccupied:

\begin{equation}
    c_i |{\psi_{\text{i unoccupied}}}\rangle = 0
\end{equation}

Note that $0$ here is not the vacuum state $|{0}\rangle$, but rather the null vector in the Hilbert space. 

Using both the antisymmeterisation condition and the action of a operator on an unoccupied state, one may derive the following commutation relations:

\begin{equation}
\left[c_{i}^{\dagger}, c_{j}^{\dagger}\right]=\left[c_{i}, c_{j}\right]=0, \quad\left[c_{i}, c_{j}^{\dagger}\right]=\delta_{i j}
\end{equation}
