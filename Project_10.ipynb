## Introduction

It is quite apparent that life without vision is not endearing. To protect the sense of vision, one has to take proper care of the organ responsible for sight, the eye. \\
The human eye consists of eight primary parts: The cornea, iris, pupil, sclera, lens, vitreous humor, retina, and optic nerve. The anteriormost part of the eye is the cornea, a transparent layer that protects the iris and the pupil. In addition to shielding, it also contributes to about 70% of the focusing of light. The stroma of the cornea accounts for about 90% of the thickness. It comprises about 250 flattened lamellae (layer of collagen fibrils), overlapping each other (1-3 μm thick and 10-100 μm wide). The anterior lamellae interweave more than the posterior lamellae. Damage to the stroma can cause irritation, but if highly damaged, it can cause blindness. Therefore, it is of great interest to develop in vivo imaging techniques that provide structural and quantitative information about the corneal stroma. \\
Multiphoton microscopy (MPM) can produce corneas images in *ex vivo* conditions via Second Harmonic Generation (SHG). However, the collagen fibrils' diameter is only about 30 nm, which is much smaller than optical resolution, and hence cannot be resolved using SHG. Hence, Backward SHG (B-SHG) images do not have enough features for clinical examination. However, Forward SHG (F-SHG) images have line-like features, which show the lamellae's orientation and can be used to describe the stroma structure. 
As discussed, SHG can be used to study the cornea in ex vivo conditions. In this paper, we shall discuss the use of B-SHG microscopy to receive the orientation information of the collagen fibrils and to enable *in vivo* imaging of the cornea.

## Theoretical Background:

Second Harmonic Generation is a particular case of Sum-Frequency Generation, where the incident frequency photons have the same frequency.  SHG is a nonlinear process in which the incident photons interact to form photons of double the frequency of incident photons. \\
The second-order nonlinear susceptibility ($\chi^{(2)}$, a tensor) of a medium obtained by averaging the collagen fibrils characterizes its tendency to cause SHG. \\
The second-order polarization induced by an electric field E in a medium is given by: $$P_i^{(2)} = \chi_{ijk}^{(2)}E_jE_k$$

Then tensor provides the orientation of these fibrils. Since the collagen lamellae are parallel to the cornea, which focuses the incident beam, it is assumed that the collagen fibrils and the incident field are both within the focal plane. Also, there is often a symmetry condition in nonlinear optics that is fulfilled when unusual dispersion can be ignored, and then the susceptibility remains unchanged when the frequencies of the input signals are variated. So the components of susceptibility tensor must be independent of all alterations. This is known as the Kleinmann symmetry condition. We assume that the Kleinmann symmetry condition is valid because of the nonresonant character of the interaction. \\
After considering the Kleinmann symmetry of collagens and the cylindrically symmetrical structure of the lamellae, the 3D tensor reduces to only two independent components: $\chi_{xxx}$ and $\chi_{xyy} = \chi_{xzz} = \chi_{yxy} = \chi_{zxz} = \chi_{yyx} = \chi_{zzx}$. \\
For a laser beam at an angle $\alpha$ to the cornea lying in X-Y plane, the induced second harmonic (SH) radiation is given by: 
$${E_x^{(2\omega)} \propto (\chi_{xxx}\cdot cos^2\alpha + \chi_{xyy} \cdot sin^2\alpha)E_0^2}$$
$$E_y^{(2\omega)} \propto (\chi_{xyy}\cdot sin (2\alpha))E_0^2 $$

and the intensity for polarization is given by:
$$I_x^{(2\omega)} = k\,|\chi_{xxx}\cdot cos^2\alpha + \chi_{xyy}\cdot sin^2 \alpha|^2$$
$$I_y^{(2\omega)} = k\,|sin (2*alpha)|^2$$

where k is a constant merging the geometry and squared incident beam intensity. \\
Taking $\rho = \frac{\chi_{xxx}}{\chi_{xyy}}$, the total SH intensity is thus given by:
$$I^{(2\omega)} = k\,(| \rho \cdot cos^2\alpha + sin^2\alpha|^2 + |sin(2\alpha)|^2)$$

For different orientations of fibrils, the above set of equations are modified:
$$I^{(2\omega)} = k\,(| \rho \cdot cos^2(\alpha - \phi) + sin^2(\alpha - \phi)|^2 + |sin \,2(\alpha-\phi)|^2)$$

where $\phi$ is the fibril orientation angle (angle of collagen axis with reference axis) \\

Using the Fourier components, the above equation can be written as \\
$$ I^{(2\omega)} = A\cdot cos \,4(\alpha - \phi) + B\cdot cos \, 2(\alpha-\phi) + C$$

where $$A = \frac{k}{2} (\frac{\rho - 1}{2})^2; B = 2k(\frac{\rho - 1}{2})(\frac{\rho + 1}{2}); C = \frac{k}{2}(\frac{\rho - 1}{2})^2 + k(\frac{\rho + 1}{2})^2$$

##Experimental Setup

#### Human Cornea Preparaion:
It is crucial to know the time of removal of the cornea from the eye, the storage area before the experiment, the kind of fluid we store it in, after what time of the removal experiment take place. These are all things we have to consider before the beginning the experiment, and have to arrange the cornea accordingly. \\
The study shown in this paper is conducted according to the tenets of the Declaration of Helsinki and French legislation for the scientific use of human corneas, using 7 human corneas, obtained from the French Eye Bank (BFY, Paris, France), which were unsuitable for transplantation. They were stored at 31°C in the CorneaMax medium (Eurobio, France) until the experiment. A custom-built holder was used for SHG imaging that clamped the sclera and preserved the cornea. This holder was inserted in a Petri dish with refined thickness to optimize the F-SHG signal detection. Corneas were immersed in a storage medium without red phenol to prevent tissue drying during SHG imaging. They were slightly edematous as shown by their increased thickness (700 µm) compared to physiological corneal thickness (around 500 µm). 

#### Imaging Setup:
The experimental setup consists of a custom-built laser scanning microscope to perform multiphoton imaging where excitation is provided by femtosecond titanium-sapphire laser and scanned in XY direction using photon-counting detection (P25PC photomultiplier tubes, Electron Tubes). To achieve lateral and axial spatial resolution near the surface, the experimentalists used a high numerical aperture water immersion objectives. Suitable filters were used to reject the excitation laser beam and to catch the 2PEF and SHG. This B-SHG and 2PEF are used for *in vivo* imaging, and F-SHG is also detected for *ex vivo* imaging of the cornea. The laser power at the objective focus is typically 20 to 85 mW. Pixels of around 0.5 to 0.8 µm size are recorded at the rate of 200 to 300 kHz to create the image. \\
A quarter waveplate and a half waveplate can be seen in the image near the objective. These achromatic waveplates are used for the Polarization resolved SHG. A quarter waveplate is used to correct the 14% ellipticity that is introduced by the galvanometric mirrors and a dichroic mirror to obtain a well-defined polarization, or it is used to switch to quasi-circular polarization. A half waveplate is used to control the orientation of this linear polarization. We make sure that forward and backward 2PEF signals from fluorescent slab exhibit variations less than 3.2% when the linear excitation is rotated. \\
By analyzing the polarization of 2PEF signals using a polarizing beamsplitter and linear polarizers in the forward detection module the absolute polarization angle of the laser excitation is verified before each experiment(Fig 1C).Equivalently,we summed all the SHG images obtained with linearly polarized excitation tuned from −90° to 90° every 10°.

## Results

#### Cornea Multiphoton Imaging:
Typical B-SHG and F-SHG images from ex vivo human corneas are displayed in following figure. We can see that F-SHG images are characterised by striated spatial features, while B-SHG are nearly homogeneous at μm scale. B-SHG does not directly provide any structural information. But after measurements we find that the intensity ratio after normalization of both detection channels of B-SHG to F-SHG is almost 1. That is, B-SHG signals are quite large signals. We can perform polarization-resolved SHG microscopy as we will get larger SHG signals when the incident excitation field aligns with the collagen fibrils. We record both F-SHG and B-SHG polarization-resolved signals by rotating incident polarization, and then process the data using nonlinear optics formalism.

#### Determination of Fibril Orientations:
Graphs plotted below are the polarimetric diagrams for the ROIs highlighted in the SHG images, where $\phi$ corresponds to the orientation of the striated features that appear in F-SHG. It confirms that striated features reveal orientation of the collagen fibrils in stromal lamellar domains. The fitting gives a value of $R^2$ close to 1. In a few ROIs $R^2$ values were lower and fitting F-SHG and B-SHG signals provided different parameters. This is due to the presence of specific structures in these ROIs. F-SHG image showing 2 distinct networks of striated features superimposed implies that the excitation beam was focused at the interface between two lamellae and SHG radiation from these adjacent lamellae were collected together. Thus, fibril orientation cannot be determined unambiguously in such ROIs.

```
import matplotlib.pyplot as plt 
import numpy as np
```

```
def funEq2(alpha, rho, phi, k):
    term1 = rho*((np.cos(alpha-phi))**2)
    term2 = (np.sin(alpha-phi))**2
    term3 = (np.sin(2*(alpha-phi)))
    return k *(((term1 + term2)**2) + (term3**2))
```

```
def funEq3(alpha, rho, phi, k):
    A = (k/2) * (((rho-1)/2)**2)
    B = (k*2) * ((rho-1)/2) * ((rho+1)/2)
    C = ((k/2) * (((rho-1)/2)**2)) + (k) * (((rho+1)/2)**2)
    return A*(np.cos(4*(alpha-phi))) + B*(np.cos(2*(alpha-phi))) + C
```

We can see (in the following figure) that the goodness of the fit gradually deteriorates with depth, which is mainly due to decrease in SHG signal. ## Discussion

####Polarization-Resolved SHG of the Corneal Stroma:
In this experiment, 30 nm-diameter collagen fibrils’ orientation is resolved in lamellar domains within corneas using polarization-resolved SHG. The major advantage of this technique is that we can apply it for both forward and backward detections. This became the open door to structural quantitative in vivo microscopy. Generally, B-SHG images of the corneal stroma do not provide any structural information about the collagen lamellae distribution because of spatial homogeneity, but F-SHG images from excised corneas exhibit striated patterns that reveal the collagen fibril distribution within the lamellae, which cannot be recorded for the in-vivo configuration. Polarization-resolved SHG takes advantage of the enhanced SHG response of collagen fibrils when aligned along the excitation electric field where F-SHG and B-SHG images are recorded for various polarization directions of the laser excitation. Suitable data processing retrieves the collagen orientation in every pixel of the SHG images. The fitting of polarimetric data is based on the tensorial formalism of second-order nonlinear optics and considers possible birefringence of the cornea, which may distort the excitation polarization. ρ is the parameter that measures the anisotropy of the SHG response in any region of interest of the cornea. It is relatively low, but it sufficient to determine orientations using Eq. (3). Excitation polarization for this process must be well defined. Therefore, it is mandatory to correct any ellipticity of the incident beam and carefully control the rotating angle of the incident polarization. Our results demonstrate that this approach is (i) reliable because the obtained orientation maps are the same for F-SHG and B-SHG images and are in good agreement with the direct visualization of lamellar domain orientations in raw F-SHG images; (ii) robust and efficient since it enables the structural characterization of the lamellar domains along with the whole thickness of the corneal tissue with good fitting accuracy (high $R^2$ value)  (iii) applicable to intact eyeballs and to in vivo imaging of rat corneas.

The susceptibility tensor $χ^{(2)}$ formally applies to a homogenous medium while the cornea is composed of stacked collagen lamellae. The SHG collected signals may have contributions from two adjacent lamellae. Specific features in the depth profiles of the parameters φ, ρ and R^2 are obtained by fitting experimental polarizationre-solved SHG signals. Interfaces between lamellae were accurately located  Due to to the steep changes of the lamellar domain orientation angle φ, decrease of R2 at the same depths. Suprisingly the ρ depth profiles also exhibited minima at the lamellae interfaces, which means that a heterogeneous SH response in the focal volume results in a lower effective ρ. Numerical simulations are in excellent agreement with these experimental results. Finally, our numerical simulations demonstrate that (i) measurement of the orientation of the collagen fibrils in the lamellar domains by polarization-resolved SHG microscopy is robust even near the interface between two lamellae where the polarization diagrams are distorted; (ii) measurement of the SHG anisotropy parameter ρ must be performed away from any interface in an homogenous region of the cornea to avoid artifacts due to heterogeneities; (iii) variation of the SHG anisotropy parameter ρ is related to heterogeneities of the fibrillar collagen organization within the focal volume. The latter consideration may be used to study pathological disorders in the cornea, which are characterized by an alteration of the fibril organization inside the lamellae. Measurements of ρ should also enable monitoring of either corneal reshaping after corneal graft or corneal wounding after any injury. \\
The advantage of this polarization-resolved microscopy is that we can do in vivo microcopy.Our data shows  that clear B-SHG images are obtained in anesthetized rats despite the vital movements and that the orientation of the collagen lamellar domains can be retrieved.

##Conclusion

In this paper, they demonstrated a new method to visualize the structural organization of the cornea without any staining and in an epi-detection configuration that is appropriate for *in vivo* imaging. This approach merges polarimetric and SHG microscopies to provide quantitative data about the main direction and the heterogeneity of the collagen fibrils distribution. It takes advantage of the high sensitivity of polarimetry for optically anisotropic structures, even below the optical resolution, such as for the 30 nm-diameter stromal collagen fibrils. Using polarization-resolved SHG, we mapped the 3D distribution of collagen lamellae at each depth in human corneas and obtained an excellent agreement with numerical simulations. We also performed in vivo SHG imaging of rat corneas and demonstrated that our approach enables structural imaging of corneal stroma without any labeling and despite the vital movements. This method opens avenues for preclinical ophthalmological studies, which require dynamic follow-up of corneal lamellar structure in a variety of injuries or pathologies. It should also find applications to in vivo diagnosis of human corneal dystrophies or to corneal healing monitoring after keratoplasty or refractive surgery. More generally, epi-detected polarization-resolved SHG imaging can provide quantitative structural information about the 3D organization of collagen fibrils within any tissue. We therefore expect that this new method will extend to other organs and become a new diagnosis tool for collagen remodeling.
