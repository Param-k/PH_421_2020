## Introduction
Spontaneous Parametric Down Conversion (SPDC) is a second order nonlinear optical process, in which a photon beam is pumped into a non-centrosymmetric crystal which spontaneously annihilates the pump photon and emits 2-photons which are entangled in space-time or equivalently in wave number and frequency. If the polarization of the emitted photons are orthogonal to each other then it is called as type-2 SPDC and if polarization is same then it is type-1 SPDC, polarization of the emitted photons depends on the crystal and its configuration. The intial photon beam is dubbed the "Pump pulse", while the entangled pairs constitute what is called as the "Signal and the Idler" pair. In this project, the nature of the down-converted photon pairs is studied when the pump is a narrow pulse. This means that the pair can only be produced when the pump pulse is inside the crystal, consequently, the pulse acts like a clock that can be used to distinguish different pairs of the down-converted photons.Since in many two-photon interference experiments, the interference occurs between the amplitudes for pairs born at different times, this distinguishability leads to a decrease in twophoton interference. In type-II SPDC this effect can be eliminanted for certain choices of crystals and pump frequencies, something that we will explicitly show later.

## Interaction Hamiltonian and State vector
Hamiltonian for **Down Conversion Process** is   

$H=H_0 + \int_{V} dV\frac{2}{3}\epsilon_{0}E_{i}E_{s}E_{p}$  

where $H_{0}$ is free part of the Hamiltonian. $\chi$ is Non Linear Susceptibility;$E_p$ is Classical Pump Electric Field, and $E_s$ & $E_i$ are quantized Signal & Idler electric fields.  The pump beam is linearly polarized and propagating in the z direction. It has a central frequency $\Omega_p$ and an envelope of arbitrary shape, $\tilde(E)_p$ . Let 

$E_p(z,t) = e^{-i \Omega_p t} \tilde{E}_p(z,t)$,

$\tilde{E}_p(z,t) = \int d\nu_p \overline{E_p}(\nu_p) e^{-i k_p (\omega_p + \nu_p)z - i\nu_p t}$

When the Signal and Idler field travel along same direction in space, i.e. Collinear Case: Collinear State Vector involving converted Signal and Idler Field is $|\Psi \rangle = |0\rangle + \sum_{k_s , k_{i}} F(k_s , k_{i})a_s^{\dagger}(k_s)a_i^{\dagger}(k_i)|0\rangle$  to the First order in Interaction Hamiltonian.
$a^{\dagger}$ is photon creation operator for Signal and Idler Photon with appropriate subscripts which act on vaccum state $|0\rangle$; $F(k_s, k_i)$ is two photon spectral function.

How do we interpret this- Well how are we defining the state $\psi$, if we examine the definition, then we are saying that the state $\psi$ is essentially the sum of the vaccum state $ |0 \rangle $ and the two creation operators $a_s(k_s)$ and $a_i(k_i)$ acting on the ground state to give us the corresponding signal and idler photon states of a fixed wave-number $k_i$ and $k_s$. But what is $F(k_s, k_i)$ doing over here? Think of this as follows, suppose one operates the creation operators corresponding to extremely large wave numbers, then what happens? Obviously, the probability for that to happen is going to be very small, since the pump is only providing finite amount of energy (This will be a case of severe phase mismatch between $k_p, k_i$ and $k_s$ but the probability can still be non-zero because of the energy-time uncertainty principle, something that we recently saw in QM2). Hence, $F(k_s, k_i)$ is essebtially capturing that probability for the biphoton pair to be of the given wave numbers!

## Biphoton amplitude for type-2 down-conversion:

In type-II SPDC the signal and idler are orthogonally polarized. We shall take the extraordinary ray (e ray) to be the
signal and the ordinary ray (o ray) to be the idler. Well the physics behind all this is what we call as Double refraction, or Birefringence. It's an optical property in which a single ray of unpolarized light entering an "anisotropic" medium is split into two rays, each travelling in a different direction. One ray, called the extraordinary ray is bent, or refracted, at an angle as it travels through the medium, while the other, called the ordinary ray passess through the medium unchanged.

In double refraction, the Ordinary and the extraordinary ray are polarized in planes vibrating in perprndicular directions to each other (Type II SPDC). The refractive index in the direction in which the orinary ray travels is constant while it changes in the direction in which the extraordinary ray travels. This property is central to what are called as "Uniaxial Crystals", which have same refractive index in two directions but a different refractive index in the third direction. That special direction/axis is what one calls as "Extraordinary axis" or the "Optical axis". What we have here is a negative Uniaxial crystal, which basically means that the refractive index in the direction in which extraordinary ray traverses is less than the one in which the ordinary ray travels.

Quantity of interest- Coincidence count: So the Physics behind this goes as follows- Not all the biphoton pairs that are produced are entangled, and hence not of interest to us. What we want, or are interested in are the entangled biphoton pairs, and coincodence count gives us the probability for the produced biphoton pair to be entangled. We define this as follows

$R_c = \int_0 ^\infty dT_1 \int_0 ^\infty dT_2 \langle \psi | E_1^{(-)} E_2^{(-)} E_2^{(+)} E_1^{(+)} | \psi \rangle$

where the  electric fields are free space fields evaluated at detectors 1 and 2. To see how this expression comes about, we refer the enthusiast reader to https://journals.aps.org/pr/abstract/10.1103/PhysRev.130.2529

The field at detector 1 can be written as

$E_1^{(+)} = \frac{1}{\sqrt2} \sum_\omega e_\omega a(k(\omega)) e^{-i \omega t_1}$

where $a(k))$  is the annihilation operator of a photon with wave number k at the output face of the crystal, $e_\omega = (\hbar \omega / \epsilon_0 V_Q)^{1/2}$, $V_Q$ is the quantization volume, $t_1 = T_1 - l_1 /c$, and $l_1$ is the optical path length from the output face of the crystal to detector 1. The field $E_2$ is defined in the same way except there is an extra factor of i multiplying the field because of the reflection from the beam splitter. We now define the following-

$\langle \psi | E_1^{(-)} E_2^{(-)} E_2^{(+)} E_1^{(+)} | \psi \rangle = |\langle 0 | E_2^{(+)} E_1^{(+)} | \psi \rangle|^2 = | A(t_+ , t_{12} )|^2 $

The quantity $A(t_+ , t_{12})$ is called as the "Biphoton" or the Two-Photon Amplitude and is defined as 

$A(t_+ , t_{12}) = \langle 0 | E_2^{(+)} E_1^{(+)} | \psi \rangle$,

where we have introduced the times-

$t_+ = \frac{1}{2} (t_1 + t_2)$,

$t_{12} = t_1 - t_2$.

But what are these $t_1, T_1, t_2$ and $T_2$? Well, the interpretation goes as follows- $t_i$ is the time at which the beam comes out from the other edge of the crystal, and $T_i$, as should be clear from the definition is the time at which the beam hits the detector. Hence, their difference is just a measure of the time of flight, so as to say. The index 1 is associated to signal (e ray), while the index 2 is associated to idler (o ray). We may think of $t_+$ as the time at which the center of momentum of the biphoton arrives at the output face of the crystal. When the frequency filters centered around $\Omega_s = \Omega_i$ are
placed in front of the detectors, we show that 

$A(t_+ , t_{12})  = \nu (t_+) u(t_{12}) w(t_+, t_{12})$.

We provide an expilicit derivation of this biphoton amplitude. Plugging in the definition of $| \psi \rangle$ in the definition of the biphoton amplitude, we get the following

$A(t_+ , t_{12}) =  \langle 0| E_2^{(+)} E_1^{(+)} | 0 \rangle + \sum_{k_s, k_i} F(k_s, k_i) \langle 0| E_2^{(+)} E_1^{(+)} a_s^\dagger(k_s) a_i^\dagger(k_i) | 0 \rangle $

We put in the definition of $E_1^{(+)}$ and $E_2^{(+)}$ again over here for reader's convenience

$E_1^{(+)} = \frac{1}{\sqrt2} \sum_\omega e_\omega a(k(\omega)) e^{-i \omega t_1}$

$E_2^{(+)} = \frac{i}{\sqrt2} \sum_\omega e_\omega a(k(\omega)) e^{-i \omega t_2}$

From this, it should be clear that the first term in the expansion for biphoton amplitude $A(t_+ , t_{12})$ trivially vanishes. The second term is non-zero, and we pick us two Dirac delta's in it's evaluation- $\delta_{w, k_sc_s}$ and $\delta_{w', k_ic_i}$. Using this, the expression for the Biphoton reduces to

$A(t_+ , t_{12}) = \frac{i}{2} \sum_{w,w'} e_w e_{w'} e^{-iwt_1} e^{-iw't_2} F(k_s(w), k_i(w'))$

Suppose the crystal is cut so that

$ \Omega_s + \Omega_i = \Omega_p $,

$K_s + K_i = K_p$,

$K_r = n_r \Omega_r / c$ ,   r = s,i,p

Now let $\omega = \Omega_i + \nu_i , \omega' = \Omega_s + \nu_s$ and $\omega_p = \Omega_p + \nu_p$ where, even for femtosecond pulses, we may assume that $|\nu_r| << \Omega_r$. Using this, to first order, we get

$k_r = K_r + \frac{\nu_r}{u_r(\Omega_r)}$

where $u_r(\Omega_r)$ is the group velocity of the r beam. We now compute $\Delta(k_s ,k_i)$ to first order,

$\Delta(k_s ,k_i)$ = $k_p - k_i - k_s$, the usual phase mismatch factor. Using the previous expression, we get

$\Delta(k_s ,k_i) = K_p - K_i - K_s + \frac{\nu_p}{u_p(\Omega_p)} - \frac{\nu_i}{u_i(\Omega_i)} -\frac{\nu_s}{u_s(\Omega_s)}$

Using the relation $K_s + K_i = K_p$ and the fact that $\nu_p = \nu_s + \nu_i$, we get the following

$\Delta(k_s ,k_i) = - \bigg( v_p D_+  +  \frac{1}{2} \nu_- D \bigg) $

where $\nu_- = \nu_i - \nu_s, D = \frac{1}{u_i(\Omega_i)} - \frac{1}{u_s(\Omega_s)}, D_+ = \frac{1}{2} \bigg( \frac{1}{u_i(\Omega_i)} +  \frac{1}{u_s(\Omega_s)} \bigg) - \frac{1}{u_p(\Omega_p)} $

We can now evaluate the Biphoton 

$A(t_+ , t_{12}) = \nu_1 e^{-i\Omega_p t} + I$

where the slowly varying quantities have been absorbed into $\nu_1$ and

$I = \int_{-L}^0 dz \int_{-\infty}^{\infty} d\nu_- \int_{-\infty}^{\infty} d\nu_p e^{-i(1/2)\nu_-(t_{12}+ Dz)} e^{-i\nu_p(t_+ + D_+ z)} \overline{E_p(\nu_p)}$

The integral over $\nu_-$ gives a Dirac delta function of argument $Dz + t_{12}$ which may be interpreted by saying that if a pair is created at the point z inside the crystal, $-L<z<0$, for $D$ positive, the idler will be detected at a time $D|z|$ after the signal. This is the case for a negative uniaxial crystal where the group velocity of the o ray is less than that of the e ray. We can now write

$A(t_+ , t_{12})  = \nu (t_+) u(t_{12}) w(t_+, t_{12})$

where 

$\nu(t) = \nu_0 e^{-i \Omega_p t}$,

$u(t) = \Pi(t)$,

$w(t, t') = \tilde{E} \bigg( 0, t-t' \frac{D_+}{D} \bigg)$,  and

$\Pi(t) = \frac{1}{DL}$ for $0<t< DL$ and 0 otherwise.

The effect of the finite pulse width of the pump is contained in w. Note that $D_+L$ is the difference in time for the center of momentum of the biphoton and the pump to cross the crystal. If we assume that the pump envelope peaks when its argument is zero, we see that the peak of the pump pulse arrives at the output face of the crystal when $t_+ = t_{12} D_+ / D$. To understand the meaning of this note that when $D_+ = 0$ the center of the pump pulse arrives at the output face of the crystal at the same time as the center of mass of the biphoton. If $D_+ \neq 0$ then the peak of the pulse arrives at a time $| t_{12} D_+ / D|$ relative to the center of momentum of the biphoton.  If this is a positive number then the peak of the pulse arrives after the center of mass of the biphoton. The magnitude of $t_{12} /D$ may be thought of as the distance from the output face of the crystal to the point at which the biphoton is born. In effect, the pump pulse provides us with a clock with a timing uncertainty equal to the pulse width that can be used to distinguish different biphotons. This leads to a decrease in the indistinguishability that can be observed in two-photon interference experiments.

Components and working:<br>

Pump Beam- Pulse of Gaussian envelope $\tilde{E_p}(0,t)=E_p e^{-t^2 / 2 \sigma^2}$, here pulse width is 2$\sigma$ i.e.   Full Width at Half Maximum.<br>

Crystal- BBO crystal of Length $L= 1mm$, BBO is a negative uniaxial crystal i.e. for BBO $n_e-n_o<0$. Hence $o$ ray will travel slower than $e$ ray in the crystal and will lag the $e$ ray (in terms of phase) at output face of the crystal by   $t_{12}$ <br>

Phase plates- Addition of phase: $'\tau'$ to $o$ ray and $'-\tau'$ to $e$ ray. It is adjusted such that its fast axis is parallel to $o$ ray because it lags $e$ ray.<br>

Beam Splitter (BS)- Polarization dependent 50-50 beam splitter. It splits the biphoton (superposition of $o$ and $e$ ray) by transmitting the $o$ ray to $D_1$ and reflecting the $e$ ray to $D_2$. <br>

Linear Polarizers- Oriented at $45^{\circ}$ relative to the incoming rays ($o$ and $e$ ray). Linear Polarizers are used to align the polarization of $o$ and $e$ rays in same direction so that they can interfere "properly" at CC.<br>

Detectors ($D_1,D_2$)- $D_1$ and $D_2$ will detect the incoming photon and note down the corresponding photon detected time as $T_1$ and $T_2$.<br>

Coincidence Count (CC)- Photons from $D_1$ and $D_2$ will interfere at CC and $R_c$ will be calculated from that interference.<br>

We know $$R_c=| A(t_+ , t_{12} )|^2 and  A(t_+ , t_{12} )=\nu (t_+) u(t_{12}) w(t_+, t_{12}) $$ <br>
But here, in our experiment $$A(t_+,t_{12})=\frac{1}{2} \nu(t_+)[u(t_{12}+\tau)w(t_+,t_{12}+\tau)-u(-(t_{12}-\tau)) w(t_+,-(t_{12}-\tau))]$$ <br>
Let's see how? Consider the biphoton (superposition of $o$ ray and $e$ ray) and assume its amplitude to be ($A$) at the output face of the crystal, when it passes through the phase plates $o$ and $e$ ray will got time shifted relative to each other<br>

When we calculate coincidence count the $\nu$ term will give a constant and the rest 2 terms will determine $R_c$ as a function of $\tau$. <br>

We will plot $R_c$ vs $\tau/DL$ for  $\sigma=10fs, 100fs$ and $1ps$<br>



```python
import numpy as np
import scipy.integrate as integrate

def erf(a):
    I=integrate.quad(lambda x: np.exp(-(x**2)),-a,a)[0]
    return 1/(np.sqrt(np.pi))*I
```

```python
import matplotlib.pyplot as plt
import numpy as np

tauDl=np.arange(0.0, 1.1, 0.1)
sigma=np.array([10, 100, 1000])
RcTHO=np.zeros((3, 11))

for ii in range(3):
    for jj in range(6):
        RcTHO[ii,jj]=1-(sigma[ii]*np.sqrt(np.pi)/(150))*erf((150/sigma[ii])*tauDl[jj])
        RcTHO[ii,5+jj]=1-(sigma[ii]*np.sqrt(np.pi)/(150))*erf((150/sigma[ii])*(1-tauDl[5+jj]))


plt.plot(tauDl,RcTHO[0,:],'r',tauDl,RcTHO[1,:],'g',tauDl,RcTHO[2,:],'b')
plt.show()
```

The $1ps$ pulse can be seen as a plane wave because its gives a 'vee' shaped dip which goes to zero for $\tau=DL/2$.
We have seen that depth increases with increase in $\sigma$. This proves that the uncertainity in clock increases with pump pulse width.

## 2-photon amplitude for type-1 down-conversion
In type-I SPDC, Idler and Signal photon are created with same Polarization. 

Since the two photon are created with same polarization, we could not use Polarizing BS which was used in Type II to separate Signal and Idler; so we cut the crystal appropriately so that the Signal and Idler beams which are being created are not emitted Collinearly and the signal and idler goes to Detector 1 and 2 respectively.   
The two-photon amplitude or Biphoton which is defined as $A(t_+,t_{12})=\langle 0| E_2^{(+)}E_1^{(+)}|\Psi \rangle$ where $t_+=\frac{1}{2}(t_1+t_2)$ and $t_{12}=t_1-t_2$. It can be proved that in type II case it comes out to be $A(t_+,t_{12})=v(t_+)w_I(t_+,t_{12})$  
$$\text{where}~~ v(t_+)=v_0 e^{-i \Omega_p t_+} ~~ \text{and}~~w_I(t_+,t_{12})= \int_{-L}^0 dz \tilde{E_p}(0,t_+ + D_+ z)\frac{\exp{\left(\frac{it_{12}^2}{4D"z}\right)}}{\sqrt{z}} $$  
The signal and idler are ordinary ray for the degenerate case; for which $D_+=\frac{1}{u_0(\Omega_p/2)}-\frac{1}{u_p(\Omega_p)}$ and the Dispersion $D"$ evaluated at $\Omega_p/2$ is $D"=\left.\frac{d^2K_0}{d \Omega^2} \right|_{\Omega=\Omega_p/2}$    
Width of the Biphoton which is created at z inside the crystal is described by $t_{12}$ oscillates and increases because of the dispersion. The width of the Biphoton at the exit face of the crystal becomes $\sqrt{4|D"z|}$  
   
Dispersion of the pump beam is neglected in the subsequent calculation (except fot the narrow pump width). Here we do the calculations and draw graphs for Gaussian Pump envelopes $\tilde{E_p}(0,t)=E_p e^{-t^2 / 2 \sigma^2}$. But we will have to take into account of the dispersion of the Biphoton in the calculation for degenerate type-I SPDC because in the Standard Hamiltonian the pair of photons which are being considered to be created at a single point z in the crystal.

 We take the signal and idler to be ordinary ray, and the pump to be extraordinary ray inside the crystal. By doing calculation, we get Biphoton to be $$A(t_+,t_{12})=\frac{1}{2}v\left(t_+-\frac{\tau}{2}\right)\left[w_I\left(t_+-\frac{\tau}{2},t_{12}-\tau\right)-w_I\left(t_+-\frac{\tau}{2},-t_{12}-\tau\right)\right]$$   
  
  The probability of a coincidence count is calculated to be $$R_c(\tau)=R_c(0)\left[1-\frac{J(\tau)}{J(0)}\right]$$
  $$\text{where}~J(\tau)=Re\int_{-\infty}^{\infty}dt_+\int_{-\infty}^{\infty}dt_{12}w_I(t_+,t_{12}-\tau)w_I^{*}(t_+,t_{12}-\tau)$$
  Which is further simplified to $$J(\tau)=\int_{-L}^0dz_1\int_{-L}^0dz_2\frac{\exp{-i\tau^2/D"\zeta}}{\sqrt{iz_-}}C_p(z_-)$$
  $$\text{where }z_-=z_1-z_2\text{ and  }C_p(z_-)=\int_{-\infty}^{\infty}dt_+\tilde{E_p}(0,t_++D_+z_-)\tilde{E_p^*}(0,t_+)$$
  $C_p(z_-)$ is the autocorrelation of the pump envelop.  
  For Gaussian Pump envelopes $\tilde{E_p}(0,t)=E_p e^{-t^2 / 2 \sigma^2}$ it becomes $$J(\tau)=J_0\int_0^1d\zeta \frac{1-\zeta}{\sqrt{\zeta}}\cos{\left(\frac{\tau^2}{D"L\zeta}-\frac{\tau}{4}\right)}e^{-(D_+L/2\sigma)^2\zeta^2}\text{ where }\zeta=z_-/L$$
  
  
  



```python
import matplotlib.pyplot as plt
import scipy.integrate as integ
import math
import numpy as num
from scipy.interpolate import BSpline,make_interp_spline
x_axis = num.arange(0.0,9,0.5)
y_axis_1 = num.arange(0.0,1.8,0.1)
y_axis_2 = num.arange(0.0,1.8,0.1)
y_axis_3 = num.arange(0.0,1.8,0.1)
one = num.zeros(300)


def f(t,d):
    return integ.quad(lambda k: (1-k)*1/(math.sqrt(k))*math.cos((t/k)-(math.pi/4) )*math.exp(-d*k**2),0,1)

for i in range(18):
    y_axis_1[i] =1-( (f(x_axis[i],5)[0])/f(0,5)[0])
    y_axis_2[i] =1-( (f(x_axis[i],1)[0])/f(0,1)[0])
    y_axis_3[i] =1-( (f(x_axis[i],0)[0])/f(0,0)[0])
for j in range(300)  :
    one[j]=1
    
xnew = num.linspace(0, 8, 300)
spl = make_interp_spline(x_axis, y_axis_1 , k=3)
y1_smooth = spl(xnew)
spl = make_interp_spline(x_axis, y_axis_2 , k=3)
y2_smooth = spl(xnew)
spl = make_interp_spline(x_axis, y_axis_3 , k=3)
y3_smooth = spl(xnew)


plt.plot(xnew,y1_smooth,linestyle='-')
plt.plot(xnew,y2_smooth,linestyle='--')
plt.plot(xnew,y3_smooth,linestyle='dotted')
plt.plot(xnew,one,color='black' ,linestyle=':')
```

In the above figure, it can be observed that the width of the dip decreases as the pulse width of pump($\sigma$) decreases.

The width of the peak on the other hand, is governed by two dimensionless quantities namely- $a=\frac{\Delta \tau^2}{D"L}$ and  $b=\frac{D_+L}{2\sigma}$. The equations governing width of the interference as measured from dip to the first maximum can then be found out(through a bit of tedious calculations) to be:
$$1/a^2=c_1b^2+c_2$$
$$\frac{1}{\Delta \tau^4}=c_1\left(\frac{D_+}{2D"}\right)^2\frac{1}{\sigma^2}+\frac{c_2}{(D"L)^2}$$
